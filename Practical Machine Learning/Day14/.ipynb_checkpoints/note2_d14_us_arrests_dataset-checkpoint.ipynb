```python
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from sklearn.metrics import silhouette_score

from scipy.cluster.hierarchy import fcluster
```

```python
df = pd.read_csv("USArrests.csv", index_col = 0)
df.head(3)
```

```python
scaler = StandardScaler().set_output(transform = "pandas")
crime_scaled = scaler.fit_transform(df)
```

```python
link = "average"
mergings =  linkage(crime_scaled, method = link)

distance_threshold = 2.4

plt.figure(figsize = (15, 8))
dendrogram(mergings, 
           labels = list(crime_scaled.index),
          color_threshold= distance_threshold
          )

plt.axhline(y=distance_threshold, color='r', linestyle='--')
plt.title(link + " linkage")
plt.xticks(rotation = 75, fontsize = 12)
plt.show()
```

```python
# With AggloAgglomerativeClustering function

clust = AgglomerativeClustering(n_clusters= 7, 
                                linkage= "average",
                               )
clust.fit(crime_scaled)
clust.labels_
```

```python
df_clust = pd.DataFrame({"Animal" : list(crime_scaled.index), "Cluster" : list(clust.labels_)})
df_clust.sort_values("Cluster")
```

```python
scores = []
n_clust = [2, 3, 4, 5, 6] 
for i in n_clust :
    clust = AgglomerativeClustering(n_clusters = i, linkage = "average")
    clust.fit(crime_scaled)
    scores.append(silhouette_score(crime_scaled, clust.labels_))
print(scores)
i_max = np.argmax(scores)
n_clust[i_max]
print("Best no. of Clusters:", n_clust[i_max])
```

# Nutrient Dataset DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
nutrient = pd.read_csv("nutrient.csv", index_col = 0)
nutrient
```

```python
scaler = StandardScaler().set_output(transform = "pandas")
nutri_scaled = scaler.fit_transform(nutrient)
```

```python
# Here we are finding the best parameters

eps_range = [0.2, 0.4, 0.6, 1]
mp_range = [2, 3, 4, 5]
cnt = 0
a = []

for i in eps_range:
    for j in mp_range:
        clust_DB = DBSCAN(eps = i, min_samples = j)
        clust_DB.fit(nutri_scaled.iloc[:,:5])
        if len(set(clust_DB.labels_)) > 2:
            cnt = cnt + 1
            nutri_scaled["Clust"] = clust_DB.labels_
            nutri_scl_inliners = nutri_scaled[nutri_scaled["Clust"]!=-1]
            sil_sc = silhouette_score(nutri_scl_inliners.iloc[:,:-1], nutri_scl_inliners.iloc[:,-1])
            a.append([cnt, i, j, sil_sc])
            print(i, j, sil_sc)


a = np.array(a)
pa = pd.DataFrame(a, columns = ['Sr', 'eps', 'min_pt', 'sil'])   
print("Best Parameters")
pa[pa['sil'] == pa['sil'].max()]
```

```python
nutri_scaled
```

```python
clust_DB = DBSCAN(eps = 0.4, min_samples = 2)
clust_DB.fit(nutri_scaled.iloc[:,:-1])
print(clust_DB.labels_)
print(len(set(clust_DB.labels_)))
```

### Cluster Analysis

```python
nutri_clust = nutrient.copy()
nutri_clust['cluster'] = clust_DB.labels_
print(clust_DB.labels_)
nutri_clust = nutri_clust[nutri_clust['cluster'] != -1]
nutri_clust.groupby('cluster').mean()
```

nutri_clust.sort_values('cluster')

# K means Clustering

```python
from sklearn.cluster import KMeans
```

```python
milk = pd.read_csv("milk.csv", index_col = 0)
scaler = StandardScaler().set_output(transform="pandas")
milkscaled = scaler.fit_transform(milk)

n_clusters = [2, 3, 4, 5]
cnt = 0
a = []

for i in n_clusters:
   
    clust_DB = KMeans(n_clusters = i, random_state =24)
    clust_DB.fit(milkscaled.iloc[:,:5])
    if len(set(clust_DB.labels_)) > 2:
        cnt = cnt + 1
        milkscaled["Clust"] = clust_DB.labels_
        milk_scl_inliners = milkscaled[milkscaled["Clust"]!=-1]
        sil_sc = silhouette_score(milk_scl_inliners.iloc[:,:-1], milk_scl_inliners.iloc[:,-1])
        a.append([cnt, i, sil_sc])
        print(i, sil_sc)


a = np.array(a)
pa = pd.DataFrame(a, columns = ['k', 'k', 'sil'])   
print("Best Parameters")
pa[pa['sil'] == pa['sil'].max()]
```

```python
clust_DB = KMeans(n_clusters = 4, random_state=24)
clust_DB.fit(milkscaled)
print(clust_DB.labels_)
print(len(set(clust_DB.labels_)))
```

```python
milk_clust = milk.copy()
milk_clust['cluster'] = clust_DB.labels_
milk_clust = milk_clust[milk_clust['cluster'] != -1]
milk_clust.groupby('cluster').mean()
```

```python
nutri_clust.sort_values('cluster')
```

# RFM Segmentation

```python
rfm = pd.read_csv("rfm_data_customer.csv", index_col = 0)
rfm = rfm.drop("most_recent_visit", axis = 1)
# rfm = rfm.sort_index()
rfm.head()
```

```python

scaler = StandardScaler().set_output(transform="pandas")
rfm_scaled = scaler.fit_transform(rfm)

n_clusters = [2, 3, 4, 5]
scores = []

for i in n_clusters:
   
    clust_DB = KMeans(n_clusters = i, random_state = 24)
    clust_DB.fit(rfm_scaled)
    sil_sc = silhouette_score(rfm_scaled, clust_DB.labels_)
    scores.append([i, sil_sc])
    print(i, sil_sc)


scores = np.array(scores)
pa = pd.DataFrame(scores, columns = ['k', 'sil'])   
print("Best Parameters")
pa[pa['sil'] == pa['sil'].max()]
```

```python
clust_DB = KMeans(n_clusters = 3, random_state=24)
clust_DB.fit(rfm_scaled)
print(clust_DB.labels_)
print(len(set(clust_DB.labels_)))
```

```python
rfm_clust = rfm.copy()
rfm_clust['cluster'] = clust_DB.labels_
rfm_clust.groupby('cluster').mean()
```

```python
rfm_clust.sort_values('cluster')
```
