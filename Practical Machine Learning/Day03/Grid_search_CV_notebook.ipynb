```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import ElasticNet
import numpy as np
import pandas as pd
```

```python
boston = pd.read_csv("Boston.csv")
X = boston.drop('medv', axis = 'columns')
y = boston['medv']
```

```python
ela = ElasticNet()
params = {'alpha' : np.linspace(0.0001, 10, 20),
          "l1_ratio" : np.linspace(0.0001, 1, 10)}
gridCV = GridSearchCV(ela, param_grid = params)
gridCV.fit(X, y)

```

```python
gridCV.best_params_
```

```python
gridCV.best_score_
```
