2. Imputation : A way of replaceing missing values in a dataset

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
chemicalProcess = pd.read_csv("ChemicalProcess.csv")
```

```python
X = chemicalProcess.drop('Yield', axis = 'columns')
y = chemicalProcess['Yield']
```

```python
chemicalProcess.head(3)
```

```python
chemicalProcess.isna().sum().sum()
```

Four strategies for imputations: 
1. mean
2. median
3. most frequent
4. constant - (customizable)

preprocessing is important as it improves the quality of data which will be used for model training

Advantage of using SimpleImputer instead of 

##### Constant Strategy

```python
impConst = SimpleImputer(
    strategy = 'constant', 
    fill_value = 0.5
).set_output(transform = 'pandas')

chemProcessConst = impConst.fit_transform(chemicalProcess.copy())
chemProcessConst.head(3)
```

##### Mean Imputation

```python
impConst = SimpleImputer(
    strategy = 'mean', 
    fill_value = 0.5
).set_output(transform = 'pandas')

chemProcessMean = impConst.fit_transform(chemicalProcess.copy())
print(chemProcessMean.isna().sum().sum())
chemProcessMean.head(3)
```

##### Median Imputation

```python
impConst = SimpleImputer(
    strategy = 'median', 
    fill_value = 0.5
).set_output(transform = 'pandas')

chemProcessMedian = impConst.fit_transform(chemicalProcess.copy())
print(chemProcessMedian.isna().sum().sum())
chemProcessMedian.head(3)
```

##### most_frequent Imputation

```python
impConst = SimpleImputer(
    strategy = 'most_frequent', 
    fill_value = 0.5
).set_output(transform = 'pandas')

chemProcessMode = impConst.fit_transform(chemicalProcess.copy())
print(chemProcessMode.isna().sum().sum())
chemProcessMode.head(3)
```

# Missing values in Train Test Splitted data.

use `.fit_transform()` or `.fit()` followed by `.transform()` on Training data\
use `.transform()` on test data

We want the characterstics from training data to be carried over to test data, as the model itself is built based on those characterstics.\
Those characterstics not being carried to test data is called **data leakage problem** and we want to avoid it.

```python
# Ideal way of imputation

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 24)

print(X_train.isna().sum().sum())
print(X_test.isna().sum().sum())
```

```python

imputer = SimpleImputer(
    strategy = 'mean', 
    fill_value = 0.5
).set_output(transform = 'pandas')

# Extract characterstics from training data
imputer.fit(X_train)

# Carrying those characterstics Into training set and test set
X_train_trn = imputer.transform(X_train)
X_test_trn = imputer.transform(X_test)

print(X_train_trn.isna().sum().sum())
print(X_test_trn.isna().sum().sum())
```

***Exercise*** 
1. compute linear regression values for mean imputation and median impuatation

```python
# for Mean
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 24)

imputer = SimpleImputer(
    strategy = 'mean', 
    fill_value = 0.5
).set_output(transform = 'pandas')

imputer.fit(X_train)

X_train_trn = imputer.transform(X_train)
X_test_trn = imputer.transform(X_test)

lr = LinearRegression()
lr.fit(X_train_trn, y_train)
y_pred = lr.predict(X_test_trn)
r2_score(y_test, y_pred)
```

```python
# for Median
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 24)

imputer = SimpleImputer(
    strategy = 'median', 
    fill_value = 0.5
).set_output(transform = 'pandas')

imputer.fit(X_train)

X_train_trn = imputer.transform(X_train)
X_test_trn = imputer.transform(X_test)

lr = LinearRegression()
lr.fit(X_train_trn, y_train)
y_pred = lr.predict(X_test_trn)
r2_score(y_test, y_pred)
```
