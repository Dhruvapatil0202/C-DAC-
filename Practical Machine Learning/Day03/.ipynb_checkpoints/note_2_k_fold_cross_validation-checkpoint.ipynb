Issues with standard holdout training method: \
the model does not get to work on the training data which might contain some unknown patterns.

solution: k fold cross validation\
problem of overfitting is reduced in general.

value of k should not go below 5 and not more than 20 in general \
the value of k depends greatly on data.

Train_test_split will not be used

```python
import pandas as pd
from sklearn.impute import SimpleImputer # Class
from sklearn.model_selection import train_test_split # function
from sklearn.linear_model import LinearRegression # Class
from sklearn.metrics import r2_score # function
from sklearn.model_selection import KFold # Class
from sklearn.model_selection import cross_val_score # function
from sklearn.linear_model import ElasticNet
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
import numpy as np
```

```python
boston = pd.read_csv("Boston.csv")
boston.head()
```

```python
X = boston.drop('medv', axis = 'columns')
y = boston['medv']
```

```python
lr = LinearRegression()

result = cross_val_score(lr, X, y)

result.mean()
```

```python
ela = ElasticNet()
res = cross_val_score(ela, X, y)
res.mean()
```

***Exercise***

Hyperparameter optimization for Elastic Net for the k folds crossvalidation cross_val_score

```python
arr1 = np.linspace(0.0001, 10, 20)
arr2 = np.linspace(0.0001, 1, 10)

score_arr = [[], [], []]
def temp(alp, l1):
    
    elas = ElasticNet(alpha = alp, l1_ratio = l1)
    
    scor = cross_val_score(elas, X, y).mean()
    
    score_arr[0].append(alp)
    score_arr[1].append(l1)
    score_arr[2].append(scor)

for i in arr1:
    for j in arr2:
        temp(i, j)
```

```python
temp = pd.DataFrame(data = {
    "alpha" : score_arr[0],
    "l1ratio" : score_arr[1],
    "r2_score" : score_arr[2]
})
temp.sort_values(by = 'r2_score', ascending = False)
```

# Grid Search CV

The very helpful tool for Hyperparameter optimization


```python
from sklearn.model_selection import GridSearchCV
```

```python
ela = ElasticNet()
params = {'alpha' : np.linspace(0.0001, 10, 20),
          "l1_ratio" : np.linspace(0.0001, 1, 10)}
gridCV = GridSearchCV(ela, param_grid = params)
gridCV.fit(X, y)

```

```python
gridCV.best_params_
```

```python
gridCV.best_score_
```
