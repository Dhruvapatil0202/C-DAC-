```python
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("BreastCancer.csv", index_col= 0 )
# df.Class = df.Class.apply(lambda x: 1 if x == "Malignant" else 0)
cls = df['Class']
df = df.drop("Class", axis = 'columns')
```

```python
scl = StandardScaler().set_output(transform = "pandas")
df_scl = scl.fit_transform(df)
df_scl
```

```python
pca_mod = PCA().set_output(transform = "pandas")
df_scl_pca = pca_mod.fit_transform(df_scl)
df_scl_pca
```

```python
df_scl_pca.var()
```

```python
df_pca = df_scl_pca.copy()
df_pca['Class'] = cls
sns.scatterplot(data = df_pca, x = 'pca0', y = 'pca1', hue = 'Class')
plt.show()
```

```python

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, log_loss, roc_auc_score
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
```

```python
cancer = pd.read_csv("BreastCancer.csv", index_col= 0 )
x = cancer.drop("Class", axis = 1)
y = cancer['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    random_state = 24,
                                                    test_size = 0.3,
                                                    stratify = y)
```

```python

lr = LogisticRegression(random_state = 24)

# n_components = number of most significant principle components to consider
pca_mod = PCA(n_components = 5).set_output(transform = "pandas")

scl = StandardScaler().set_output(transform = "pandas")

pipe_pca = Pipeline([("SCL", scl), ("PCA", pca_mod)])
x_trn_pca = pipe_pca.fit_transform(x_train)

lr.fit(x_trn_pca, y_train)
x_tst_pca = pipe_pca.transform(x_test)
y_pred = lr.predict(x_tst_pca)
print(accuracy_score(y_test, y_pred))
```

```python
lr = LogisticRegression(random_state = 24)


# n_components = number of most significant principle components to consider
pca_mod = PCA().set_output(transform = "pandas")
n_comps = [2,3,4,5,6]

scl = StandardScaler().set_output(transform = "pandas")

kfold = StratifiedKFold(n_splits=5, 
                        random_state=24, 
                        shuffle=True)

pipe_pca = Pipeline([("SCL", scl), ("PCA", pca_mod), ("LR" ,lr)])

params = {"PCA__n_components":n_comps,
         'LR__solver':['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga'],
         'LR__C':np.linspace(0.001, 10, 20)}

gcv = GridSearchCV(pipe_pca,
                  param_grid=params,
                scoring = "roc_auc",
                  cv = kfold,
                  verbose=3)

gcv.fit(x_train, y_train)
```

```python
y_prid = gcv.predict(x_test)
```

```python
gcv.best_score_
```

```python
gcv.best_params_
```

# Linear Discriminant Transformation

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
cancer = pd.read_csv("BreastCancer.csv", index_col= 0 )
x = cancer.drop("Class", axis = 1)
y = cancer['Class']

x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    random_state = 24,
                                                    test_size = 0.3,
                                                    stratify = y)
```

```python
lr = LogisticRegression(random_state = 24)


# n_components = number of most significant principle components to consider

lda = LinearDiscriminantAnalysis()

scl = StandardScaler().set_output(transform = "pandas")

kfold = StratifiedKFold(n_splits=5, 
                        random_state=24, 
                        shuffle=True)

pipe_pca = Pipeline([("SCL", scl), ("LDA", lda), ("LR" ,lr)])

params = {"LDA__solver" : ['svd', 'lsqr', 'eigen'],
         'LR__solver':['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga'],
         'LR__C':np.linspace(0.0000001, 10, 20)}

gcv = GridSearchCV(pipe_pca,
                  param_grid=params,
                  scoring = "roc_auc",
                  cv = kfold,
                  verbose=3)

gcv.fit(x, y)
```

```python
y_prid = gcv.predict(x_test)
```
