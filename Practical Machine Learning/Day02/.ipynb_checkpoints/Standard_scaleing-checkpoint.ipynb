```python
import numpy as np
```

```python
# Standard Scaling
```

```python
a = np.array([100, 121, 144, 169, 196, 225, 256])
b = np.array([20, 22, 24, 26, 28, 30, 32])
```

```python
ma, mb = a.mean(), b.mean()
sa, sb = a.std(), b.std()
scl_a = (a - ma) / sa
scl_b = (b - mb) / sb
```

```python
scl_a.mean()
```

```python
scl_b
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
scl_std = StandardScaler()
df = pd.DataFrame({'a': a, 'b':b})
scl_std.fit(df)
```

```python
scl_std.mean_
```

```python
scl_std.transform(df)
```

```python
# Min Max scaling

# Nearly all algorithms in neural netwroks use Min_Max scaling
```

```python
a = np.array([100, 121, 144, 169, 196, 225, 256])
b = np.array([20, 22, 24, 26, 28, 30, 32])
```

```python
# Manual calculations

min_a, min_b = a.min(), b.min()
max_a, max_b = a.max(), b.max()

a_Scaled = (a - min_a) / (max_a - min_a)
b_Scaled = (b - min_b) / (max_b - min_b)
```

```python
a_Scaled, b_Scaled

```
