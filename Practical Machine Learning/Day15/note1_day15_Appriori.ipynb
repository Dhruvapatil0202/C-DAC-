```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
fp_df = pd.read_csv("Faceplate.csv", index_col = 0)
fp_df = fp_df.astype(bool)
fp_df
```

```python
itemsets = apriori(fp_df, min_support = 0.2, use_colnames = True)
rules = association_rules(itemsets, metric='lift', 
                          min_threshold=0.3, 
                          num_itemsets= len(fp_df),
                         return_metrics = ['support', 'confidence', 'lift'])
rules.sort_values('confidence', ascending = False)
```

# Groceries

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
groceries = []
with open("Groceries.csv", "r") as f:
    groceries = f.read()

groceries = groceries.split("\n")

groceries_list = []

for i in groceries:
    groceries_list.append(i.split(","))

te = TransactionEncoder().set_output(transform = "pandas")
te_ary = te.fit_transform(groceries_list)
te_ary
```

```python
itemsets = apriori(te_ary, min_support = 0.01, use_colnames = True)
# print(itemsets)

rules = association_rules(itemsets, 
                          metric = 'lift', 
                          min_threshold = 1, 
                          num_itemsets= len(te_ary),
                         return_metrics = ['support', 'confidence', 'lift'])

rules.sort_values('confidence', ascending = False)
```
