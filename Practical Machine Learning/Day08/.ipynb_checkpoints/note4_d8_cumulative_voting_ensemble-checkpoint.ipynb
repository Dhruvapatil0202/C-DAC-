# Cancer dataset

```python
df = pd.read_csv("BreastCancer.csv", index_col = 0)
df
```

```python
x, y = df.drop("Class", axis = 1), df.Class
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.2,
                                                    random_state = 24,
                                                    stratify = y)
```

```python
# -------------------------------------------------------------------------------
# Voting Classifier

dtc = DecisionTreeClassifier(random_state = 24)
lr = LogisticRegression(random_state = 24)
nb = GaussianNB()


voting = VotingClassifier(
    [("DT", dtc), ("LR", lr), ("NB", nb),],
    voting = "soft",
    )

# -------------------------------------------------------------------------------
# Params

params = {
    "DT__max_depth" : [None, 4, 6],
    "DT__min_samples_leaf" : [1, 10, 20],
    "DT__min_samples_split" : [2, 10, 20],
    "LR__C" : np.linspace(0.001, 3, 5),
}

kfolds = StratifiedKFold(n_splits = 5,
                        random_state = 24,
                        shuffle = True)

gcv = GridSearchCV(voting,
                  param_grid = params,
                  scoring = "roc_auc",
                  cv = kfolds,
                  verbose = 3)

gcv.fit(x, y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
voting.fit(x_train, y_train)
y_pred = voting.predict(x_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
y_pred = voting.predict(x_test)
y_pred_prob = voting.predict_proba(x_test)[:, 1]
print(roc_auc_score(y_test, y_pred_prob))
```

```python
# ______________________________________________________________________________________
# Pipeline

# one hot encoder
ohe = OneHotEncoder(
    handle_unknown="ignore",
    sparse_output=False,
    drop='first'
).set_output(transform='pandas')

trans_ohe = make_column_transformer(
    ('passthrough', make_column_selector(dtype_exclude=object)),
    (ohe, make_column_selector(dtype_include=object)),
    verbose_feature_names_out=False
).set_output(transform='pandas')


# scaler
scl_std = StandardScaler().set_output(transform = "pandas")
scl_mm = MinMaxScaler().set_output(transform = "pandas")


# model
# dtc = DecisionTreeClassifier(
#     random_state = 24,
#     # max_depth = 4
# )
lr = LogisticRegression(
    random_state = 24,
)


pipe_pca = Pipeline([("OHE", trans_ohe), ("SCL", scl_std), ("LR", lr)])


# _____________________________________________________________________________________
# GCV

params = {
    "DTC__max_depth" : [None, 4, 6],
    "DTC__min_samples_leaf" : [1, 10, 20],
    "DTC__min_samples_split" : [2, 10, 20],
}

kfolds = StratifiedKFold(n_splits = 5,
                        random_state = 24,
                        shuffle = True)

gcv = GridSearchCV(pipe_pca,
                  param_grid = params,
                  scoring = "roc_auc",
                  cv = kfolds,
                  verbose = 3)

gcv.fit(x, y)
# gcv.fit(x, y)
```

# Concrete Strength (Voting Regressor)

```python
df = pd.read_csv("Concrete_Data.csv")
df
```

```python
x, y = df.drop("Strength", axis = 1), df.Strength
```

```python
# x_test = pd.read_csv("testConcrete.csv")
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    random_state = 24,
                                                    test_size = 0.3)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
r2_lr = r2_score(y_test, y_pred)
r2_lr
```

```python
dt = DecisionTreeRegressor(random_state = 24)
dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
r2_dt = r2_score(y_test, y_pred)
r2_dt
```

```python
eln = ElasticNet()
eln.fit(x_train, y_train)
y_pred = eln.predict(x_test)
r2_eln = r2_score(y_test, y_pred)
r2_eln
```

```python
# ______________________________________________________________________________________
# Pipeline

# one hot encoder
ohe = OneHotEncoder(
    handle_unknown = "ignore",
    sparse_output=False,
    # drop='first'
).set_output(transform='pandas')

trans_ohe = make_column_transformer(
    ('passthrough', make_column_selector(dtype_exclude=object)),
    (ohe, make_column_selector(dtype_include=object)),
    verbose_feature_names_out=False
).set_output(transform='pandas')


# scaler
scl_std = StandardScaler().set_output(transform = "pandas")
scl_mm = MinMaxScaler().set_output(transform = "pandas")


# model
# dtc = DecisionTreeClassifier(
#     random_state = 24,
#     # max_depth = 4
# )

# Voting Classifier

lr = LinearRegression()
dt = DecisionTreeRegressor(random_state = 24)
eln = ElasticNet()


voting = VotingRegressor(
    [ ("LR", lr), ("DT", dt), ("ELN", eln),],
    # voting = "soft",
    weights = [r2_lr, r2_dt, r2_eln]
    )

voting.fit(x_train, y_train)
y_pred = voting.predict(x_test)
r2_score(y_test, y_pred)

pipe_pca = Pipeline([("OHE", trans_ohe), ("SCL", scl_std), ("VOT", voting)])

# _____________________________________________________________________________________
# GCV

params = {
    "VOT__DT__max_depth" : [None, 4, 6],
    "VOT__DT__min_samples_leaf" : [1, 10, 20],
    "VOT__DT__min_samples_split" : [2, 10, 20],
}

kfolds = StratifiedKFold(n_splits = 5,
                        random_state = 24,
                        shuffle = True)

kfold = KFold(n_splits = 5,
              random_state = 24,
              shuffle = True)

gcv = GridSearchCV(pipe_pca,
                  param_grid = params,
                  scoring = "r2",
                  cv = kfold,
                  verbose = 3)

gcv.fit(x_train, y_train)
# gcv.fit(x, y)
```

```python
gcv.best_score_
```

```python
dtr1 = DecisionTreeRegressor(random_state = 24, max_depth = None)
dtr2 = DecisionTreeRegressor(random_state = 24, max_depth = 3)
dtr3 = DecisionTreeRegressor(random_state = 24, max_depth = 4)
voting = VotingRegressor([("m1", dtr1), ('m2', dtr2), ('m3', dtr3)],
                         weights = [0.7, 0.15, 0.15])
voting.fit(x_train, y_train)
y_pred = voting.predict(x_test)
r2_score(y_test, y_pred)
```
