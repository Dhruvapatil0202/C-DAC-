# AdaBoosting

```python
df = pd.read_csv("Breastcancer.csv")
df.head(3)
```

```python
x, y = df.drop("Class", axis = 1), df["Class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    random_state = 24, 
                                                    test_size= 0.3,
                                                    stratify= y)
```

```python
ada = AdaBoostClassifier(
    n_estimators= 155,
    random_state= 24,
)

ada.fit(x_train, y_train)
y_pred= ada.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python


# ______________________________________________________________________________________
# Pipeline

# one hot encoder
ohe = OneHotEncoder(
    handle_unknown="ignore",
    sparse_output=False,
    drop='first'
).set_output(transform='pandas')

trans_ohe = make_column_transformer(
    ('passthrough', make_column_selector(dtype_exclude=['category', object])),
    (ohe, make_column_selector(dtype_include=['category', object])),
    verbose_feature_names_out=False
).set_output(transform='pandas')


# scaler
scl_std = StandardScaler().set_output(transform = "pandas")
scl_mm = MinMaxScaler().set_output(transform = "pandas")


# Model

dtc = DecisionTreeClassifier(
    random_state = 24
)

adac = AdaBoostClassifier(
    estimator = dtc,
    random_state = 24
)

pipe = Pipeline([("ADAC", adac)])


# _____________________________________________________________________________________
# GCV

params = {
    "ADAC__n_estimators" : [10, 50, 100, 150, 155],
    "ADAC__estimator__max_depth" : [1, 2, 3]
}

kfolds = StratifiedKFold(n_splits = 5,
                        random_state = 24,
                        shuffle = True)

kfold = KFold(n_splits = 5,
            random_state = 24,
            shuffle = True)

gcv = GridSearchCV(pipe,
                  param_grid = params,
                  scoring = "neg_log_loss",
                  cv = kfolds,
                  verbose = 3)


# gcv.fit(x, y)
```

```python
pipe.get_params()
```

```python
gcv.fit(x, y)
```

```python
print(gcv.best_score_)
print(gcv.best_params_)
```

```python
y_pred_prob = gcv.predict_proba(x_test)[:, 1] 
print(log_loss(y_test, y_pred_prob))
```

# Gradient Boosting

```python
df = pd.read_csv("Breastcancer.csv")
df.head(3)
```

```python
x, y = df.drop("Class", axis = 1), df["Class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    random_state = 24,
                                                    test_size = 0.3,
                                                    stratify = y)
```

```python


# ______________________________________________________________________________________
# Pipeline

# one hot encoder
ohe = OneHotEncoder(
    handle_unknown="ignore",
    sparse_output=False,
    drop='first'
).set_output(transform='pandas')

trans_ohe = make_column_transformer(
    ('passthrough', make_column_selector(dtype_exclude=['category', object])),
    (ohe, make_column_selector(dtype_include=['category', object])),
    verbose_feature_names_out=False
).set_output(transform='pandas')


# scaler
scl_std = StandardScaler().set_output(transform = "pandas")
scl_mm = MinMaxScaler().set_output(transform = "pandas")


# Model

dtc = DecisionTreeClassifier(
    random_state = 24
)


gbc = GradientBoostingClassifier(
    # estimator = dtc, 
    random_state = 24
)

adac = AdaBoostClassifier(
    estimator = dtc,
    random_state = 24
)

pipe = Pipeline([("GBC", gbc)])


# _____________________________________________________________________________________
# GCV

params = {
    "GBC__n_estimators" : list(range(30, 41)),
    "GBC__max_depth" : [2, 3, 4],
    "GBC__learning_rate" : np.linspace(0.001, 1, 5),
}

kfolds = StratifiedKFold(n_splits = 5,
                        random_state = 24,
                        shuffle = True)

kfold = KFold(n_splits = 5,
            random_state = 24,
            shuffle = True)

gcv = GridSearchCV(pipe,
                  param_grid = params,
                  scoring = "neg_log_loss",
                  cv = kfolds,
                  verbose = 3)


# gcv.fit(x, y)
```

```python
gcv.fit(x_train, y_train)
```

```python
print(gcv.best_score_)
print(gcv.best_params_)
```

```python
y_pred_prob = gcv.predict_proba(x_test)[:, 1] 
print(roc_auc_score(y_test, y_pred_prob))
```
