```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
milkProd = pd.read_csv("monthly-milk-production-pounds-p.csv", index_col = 0)
milkProd
```

```python
milkProd.plot()
```

```python
y = milkProd["Milk"]
y_train = y[:-12]
y_test = y[-12:]
y_train.shape, y_test.shape
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
from sklearn.metrics import root_mean_squared_error
from statsmodels.tsa.api import Holt
import matplotlib.pyplot as plt
```

```python
y_train.index = pd.to_datetime(y_train.index) 
y_train.index = y_train.index.to_period('M').to_timestamp()  

y_test.index = pd.to_datetime(y_test.index)  
y_test.index = y_test.index.to_period('M').to_timestamp()
```

```python
# Multiplictive Damped Trend

alpha, beta, gamma = 0.45, 0.042, 0.47

hw_a = ExponentialSmoothing(
    y_train,
    seasonal_periods= 12,
    trend = "add",
    seasonal= "mul",
    damped_trend = True
)

fit = hw_a.fit(
    smoothing_level = alpha, 
    smoothing_trend = beta,
    smoothing_seasonal = gamma,
)

fcast = fit.forecast(len(y_test))

holt_fcast = pd.Series(fcast, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

# plt.figure(figsize = (15, 8))
plt.text(2000, 1000, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "HW damped mul Forecast")
plt.plot(y_test, label = "Test")
plt.legend(loc = "best")
plt.show()
```

### Yearly Resampling

```python
milkProd = pd.read_csv("monthly-milk-production-pounds-p.csv", index_col = 0)

milkProd.index = pd.to_datetime(milkProd.index)

milkProd = milkProd.resample("YE").sum()
milkProd.plot()
```

```python
y = milkProd["Milk"]
y_train = y[:-3]
y_test = y[-3:]
y_train.shape, y_test.shape
```

```python
# Multiplictive Damped Trend

alpha, beta, gamma = 0.05, 0.06, 0.2

hw_a = ExponentialSmoothing(
    y_train,
    seasonal_periods= 12,
    trend = "add",
    # seasonal= "mul",
    damped_trend = True
)

fit = hw_a.fit(
    smoothing_level = alpha, 
    smoothing_trend = beta,
    smoothing_seasonal = gamma,
)

fcast = fit.forecast(len(y_test))

holt_fcast = pd.Series(fcast, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

plt.text(500, 11500, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "HW mul Forecast")
plt.plot(y_test, label = "Test")
plt.plot(y_train, label = "Train")
plt.legend(loc = "best")
plt.show()
```

```python
# Multiplictive Damped Trend

alpha, beta, phi = 0.1, 0.33, 0.9999

hw_a = Holt(
    y_train,
    # seasonal_periods= 12,
    # exponential = True,
    # seasonal= "mul",
    damped_trend = True
)

fit = hw_a.fit(
    smoothing_level = alpha, 
    smoothing_trend = beta,
    # smoothing_seasonal = gamma,
    damping_trend = phi
)

fcast = fit.forecast(len(y_test))

holt_fcast = pd.Series(fcast, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

plt.text(500, 11500, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "HW mul Forecast")
plt.plot(y_test, label = "Test")
plt.plot(y_train, label = "Train")
plt.legend(loc = "best")
plt.show()
```
