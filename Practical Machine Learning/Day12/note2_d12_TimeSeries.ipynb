```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import root_mean_squared_error
```

```python
passengers = pd.read_csv("AirPassengers.csv", index_col = 0)
y = passengers['Passengers']
```

```python
y_train = y[:132]
y_test = y[132:]
y_train.shape, y_test.shape
```

```python
span = 3              # window size
fcast = y_train.rolling(span, center=False).mean()
```

```python
MA = fcast.iloc[-1]
MA_fcast = pd.Series(MA.repeat(len(y_test)), index = y_test.index)
plt.plot(y_train, label = 'Data')
plt.plot(fcast, label='Rolling Mean(Train)')
plt.plot(MA_fcast, label='Rolling Mean Forcast')
plt.plot(y_test, label = 'Test')
plt.legend(loc = 'best')
plt.show()
```

```python
span = 7
fcast = y_train.rolling(span, center=False).mean()
MA = fcast.iloc[-1]
MA_fcast = pd.Series(MA.repeat(len(y_test)), index = y_test.index)
root_mean_squared_error(y_test, MA_fcast)
```

```python
plt.plot(MA_fcast, label='Rolling Mean Forcast')
plt.plot(y_test, label = 'Test')
plt.legend(loc = 'best')
plt.show()
```

## Simple Exponential Smoothing

```python
from statsmodels.tsa.api import SimpleExpSmoothing
alpha = 0.3                     # 0.3
ses = SimpleExpSmoothing(y_train)
fit1 = ses.fit(smoothing_level = alpha)
fcast1 = fit1.forecast(len(y_test))
```

```python
ses_fcast = pd.Series(fcast1, index = y_test.index)
error = round(root_mean_squared_error(y_test, ses_fcast), 2)
plt.text(1, 500, "RMSE="+str(error))
plt.plot(ses_fcast, label = 'SES Forecast')
plt.plot(y_test, label = 'Test')
plt.legend(loc = 'best')
plt.show()
```

## Holt's Linear Trend

```python
from statsmodels.tsa.api import Holt
```

```python
holt = Holt(y_train)
alpha, beta = 0.2, 0.4
fit1 = holt.fit(smoothing_level = alpha, smoothing_trend = beta)
fcast1 = fit1.forecast(len(y_test))
```

```python
holt_fcast = pd.Series(fcast1, index = y_test.index)
error = round(root_mean_squared_error(y_test, holt_fcast), 2)
plt.text(1, 500, "RMSE="+str(error))
plt.plot(holt_fcast, label = 'Holt Forecast')
plt.plot(y_test, label = 'Test')
plt.legend(loc = 'best')
plt.show()
```

### Holt's Exponential Trend

```python
holt = Holt(
    y_train,
    exponential= True
)

alpha, beta = 0.2, 0.4

fit1 = holt.fit(
    smoothing_level = alpha, 
    smoothing_trend = beta
)

fcast1 = fit1.forecast(len(y_test))

holt_fcast = pd.Series(fcast1, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

plt.text(1, 500, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "Exp Trend Forecast")
plt.plot(y_test, label = "y_test")
plt.legend(loc = "best")
plt.show()
```

Damping Trends 
1. Additive Damped Trends
2. Multiplictive Damped Trends



```python
# Additive Damped Trend
alpha, beta, phi = 0.8, 0.02, 0.1

add_damp = Holt(
    y_train,
    damped_trend= True
)

fit1 = add_damp.fit(
    smoothing_level = alpha, 
    smoothing_trend = beta,
    damping_trend= phi,
)

fcast1 = fit1.forecast(len(y_test))

holt_fcast = pd.Series(fcast1, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

plt.text(1, 500, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "Add Damped Forecast")
plt.plot(y_test, label = "Test")
plt.legend(loc = "best")
plt.show()
```

```python
# Multiplictive Damped Trend

alpha, beta, phi = 0.8, 0.02, 0.1

add_damp = Holt(
    y_train,
    damped_trend= True,
    exponential = True
)

fit1 = add_damp.fit(
    smoothing_level = alpha, 
    smoothing_trend = beta,
    damping_trend= phi,
)

fcast1 = fit1.forecast(len(y_test))

holt_fcast = pd.Series(fcast1, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

plt.text(1, 500, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "Add Damped Forecast")
plt.plot(y_test, label = "Test")
plt.legend(loc = "best")
plt.show()
```

## Holt - Winter's Method

this method comprises of forecast equation and three smoothing equations each for level, trend and seasonal component \
we us m to denote the period of season \



### Holt Winter's additive trend method

```python
from statsmodels.tsa.api import ExponentialSmoothing
```

```python
# Multiplictive Damped Trend

alpha, beta, gamma = 0.20, 0.05, 0.47

hw_a = ExponentialSmoothing(
    y_train,
    seasonal_periods= 12,
    trend = "add",
    seasonal= "add"
)

fit = hw_a.fit(
    # smoothing_level = alpha, 
    # smoothing_trend = beta,
    # smoothing_seasonal = gamma,
)

fcast = fit.forecast(len(y_test))

holt_fcast = pd.Series(fcast, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

plt.text(1, 500, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "Add Damped Forecast")
plt.plot(y_test, label = "Test")
plt.legend(loc = "best")
plt.show()
```

### Holt Winter's multiplictive trend method

Trend remains addetive and Seasonal component gets multiplied

```python
# Multiplictive Damped Trend

alpha, beta, gamma = 0.20, 0.05, 0.47

hw_a = ExponentialSmoothing(
    y_train,
    seasonal_periods= 12,
    trend = "add",
    seasonal= "mul"
)

fit = hw_a.fit(
    smoothing_level = alpha, 
    smoothing_trend = beta,
    smoothing_seasonal = gamma,
)

fcast = fit.forecast(len(y_test))

holt_fcast = pd.Series(fcast, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

plt.text(1, 500, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "HW mul Forecast")
plt.plot(y_test, label = "Test")
plt.legend(loc = "best")
plt.show()
```

### Damped + Holt-Winters

```python
# Multiplictive Damped Trend

alpha, beta, gamma = 0.25, 0.05, 0.49

hw_a = ExponentialSmoothing(
    y_train,
    seasonal_periods= 12,
    trend = "add",
    seasonal= "mul",
    damped_trend = True
)

fit = hw_a.fit(
    smoothing_level = alpha, 
    smoothing_trend = beta,
    smoothing_seasonal = gamma,
)

fcast = fit.forecast(len(y_test))

holt_fcast = pd.Series(fcast, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

plt.text(1, 500, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "HW mul Forecast")
plt.plot(y_test, label = "Test")
plt.legend(loc = "best")
plt.show()
```
