```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
beerProd = pd.read_csv("monthly-beer-production-in-austr.csv", index_col = 0)
# beerProd.index = pd.to_datetime(beerProd.index)

beerProd.index = pd.to_datetime(beerProd.index)  # Convert to DateTimeIndex (if not already)
beerProd.index = beerProd.index.to_period('M').to_timestamp()

beerProd
```

```python
y = beerProd["Monthly beer production"]
y_train = y[:-24]
y_test = y[-24:]
y_train.shape, y_test.shape
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
from sklearn.metrics import root_mean_squared_error
from statsmodels.tsa.api import Holt
import matplotlib.pyplot as plt
```

```python

# Multiplictive Damped Trend

alpha, beta, gamma = 0.10, 0.20, 0.1

hw_a = ExponentialSmoothing(
    y_train,
    seasonal_periods= 12,
    trend = "add",
    seasonal= "mul",
    damped_trend = True
)

fit = hw_a.fit(
    # smoothing_level = alpha, 
    # smoothing_trend = beta,
    # smoothing_seasonal = gamma,
)

fcast = fit.forecast(len(y_test))

holt_fcast = pd.Series(fcast, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

plt.figure(figsize = (15, 8))
plt.text(9000, 100, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "HW mul Forecast")
plt.plot(y_test, label = "Test")
# plt.plot(y_train, label = "Train")
plt.legend(loc = "best")
plt.show()
```

```python
# Multiplictive Damped Trend

alpha, beta, phi = 0.1, 0.5, 0.2

hw_a = Holt(
    y_train,
    exponential = True,
    damped_trend = True
)

fit = hw_a.fit(
    smoothing_level = alpha, 
    smoothing_trend = beta,
    damping_trend = phi
)

fcast = fit.forecast(len(y_test))

holt_fcast = pd.Series(fcast, 
                       index = y_test.index)

error = round(root_mean_squared_error(y_test, holt_fcast), 2)

plt.figure(figsize = (15, 8))
plt.text(9000, 100, "RMSE=" + str(error))
plt.plot(holt_fcast, label = "HW mul Forecast")
plt.plot(y_test, label = "Test")
# plt.plot(y_train, label = "Train")
plt.legend(loc = "best")
plt.show()
```
