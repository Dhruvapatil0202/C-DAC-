```python
smoker = pd.read_csv("smoker_train.csv", index_col = 0)
smoker.info()
```

```python
x_train = smoker.drop("smoking", axis = 1)
x_train
```

```python
y_train = smoker["smoking"]
y_train
```

```python
x_test = pd.read_csv("smoker_test.csv", index_col = 0)
x_test
```

```python


# ______________________________________________________________________________________

# scaler
scl_std = StandardScaler().set_output(transform = "pandas")
scl_mm = MinMaxScaler().set_output(transform = "pandas")


# Model

lr = LogisticRegression()


bagg = BaggingRegressor(
    estimator = lr, 
    random_state = 24)


pipe = Pipeline([("SCL", scl_mm), ("BAG", bagg)])


# _____________________________________________________________________________________
# GCV

params = {
    "BAG__n_estimators" : [2, 5, 10],
    'BAG__estimator__l1_ratio' : np.linspace(0.0001, 1, 3),
    'BAG__estimator__C' : np.linspace(0.0001, 5, 3)
}


kfolds = StratifiedKFold(n_splits = 5,
                        random_state = 24,
                        shuffle = True)

kfold = KFold(n_splits = 5,
            random_state = 24,
            shuffle = True)

gcv = GridSearchCV(pipe,
                  param_grid = params,
                  scoring = "roc_auc",
                  cv = kfold,
                  verbose = 3)


# gcv.fit(x, y)
```

```python
gcv.fit(x_train, y_train)
```

```python
gcv.best_score_
```
