```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("country_vaccinations.csv")
df
```

```python
df["people_vaccinated"] = df["people_vaccinated"].fillna(0)
df["total_vaccinations"] = df["total_vaccinations"].fillna(0)
df
```

```python
df["double_vaccinations"] = df["total_vaccinations"] - df["people_vaccinated"]
df
```

```python
df["daily_vaccinations"] = df["daily_vaccinations"].fillna(df["daily_vaccinations"].mean())
df
```

```python
df = df.drop(columns = ["source_name"], axis = 1)
df
```

```python
df = df.sort_values(by = ["total_vaccinations"],ascending = False)
df
```

```python
df = df.sort_values(by = ["country", "people_vaccinated"])
df
```

```python
df = df.rename(columns = {"iso_code":"country_code"})
df
```

```python
df.index = range(1, len(df) + 1)
df
```

```python
df["people_fully_vaccinated"] = df["people_fully_vaccinated"].fillna(0)
df
```

```python
df = df.dropna(subset=["total_vaccinations"],axis=0)
df
```

```python
df.loc[df["daily_vaccinations"] < 0, "daily_vaccinations"] = 0
df
```

```python

for index, row in df.iterrows():
    if "Moderna" in df.at[index, "vaccines"]:
        df.at[index, "vaccines"] = df.at[index, "vaccines"].replace("Moderna", "mRNA vaccines")
    if "Pfizer/BioNTech" in df.at[index, "vaccines"]:
        df.at[index, "vaccines"] = df.at[index, "vaccines"].replace("Pfizer/BioNTech", "mRNA vaccines")
df
```
