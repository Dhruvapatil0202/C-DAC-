# Data Visualization

simplifies complex data, reveals hidden insights, facilitates decision making

```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

## Matplotlib

steps to create a graph:
1. prepare the data
2. creat a plot
3. customize the plot
4. display or save the plot
   
graph types:
1. line plots
2. bar charts
3. scatter plots
4. histograph
5. pie charts

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,11)
y1 = np.array([1,3,4,7,9,12,15,18,19,22])
y2 = np.array([2,5,7,10,12,13,16,18,21,23])

plt.figure(figsize=(8,6))
plt.plot(x,y1,label = 'Dataset1', color ='blue', marker='o', linestyle='--')
plt.plot(x,y2,label = 'Dataset2', color ='red', marker='s', linestyle='-')

plt.title("Line plot example")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories = ["Category 1", "Category 2","Category 3","Category 4"]
values = [4,7,1,9]

plt.figure(figsize=(8,6))
plt.bar(categories, values, color='green')
plt.title("Bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.figure(figsize=(8,10))
plt.scatter(x,y, c=colors, s=sizes, alpha=0.5, cmap='Blues')
plt.title("Scatter plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Hands on
##### plot sales data form a year. use matplotlib for creating line plots and ba rchargs and pie charts.

```python
# Prepareing synthetic data

months = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sept", "oct", "nov", "dec"]
cols = ["product_" + str(i) for i in range(1, 6)]
prod_1_col = np.random.randint(50, 500, 12)
prod_2_col = np.random.randint(50, 500, 12)
prod_3_col = np.random.randint(50, 500, 12)
prod_4_col = np.random.randint(50, 500, 12)
prod_5_col = np.random.randint(50, 500, 12)
```

```python
df = pd.DataFrame(data = {
    'month' : months,
    'prod1' : prod_1_col,
    'prod2' : prod_2_col,
    'prod3' : prod_3_col,
    'prod4' : prod_4_col,
    'prod5' : prod_5_col,
})
df
```

```python
# Line plot

plt.figure(figsize = (15, 8))

plt.plot(df['month'], df["prod1"], label = "Product-1", color = "green", linestyle = '-')
plt.plot(df['month'], df["prod2"], label = "Product-2", color = "orange", linestyle = '-')
plt.plot(df['month'], df["prod3"], label = "Product-3", color = "lime", linestyle = '-')
plt.plot(df['month'], df["prod4"], label = "Product-4", color = "blue", linestyle = '-')
plt.plot(df['month'], df["prod5"], label = "Product-5", color = "purple", linestyle = '-')

plt.title("Sales Plot")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (15,8))
categories = months
values = [i + j + k + l + m for i,j,k,l,m in zip(prod_1_col, prod_2_col, prod_3_col, prod_4_col, prod_5_col)]
plt.bar(categories, values, color = 'blue')
plt.title("Bar chart")
plt.xlabel("months")
plt.ylabel("sales")
plt.show()
```

```python
plt.figure(figsize = (10, 10))
vals = [df[i].sum() for i in df.columns][1:]
vals
plt.pie(vals, labels = list(df.columns[1:]))
# df.columns[1:]
# plt.pie(values, months)
# list(df.columns[1:]), vals
```

```python
# pie chart
vals = [df[i].sum() for i in df.columns][1:]

axs[1, 1].pie(vals, labels = list(df.columns[1:]))
axs[1, 1].set_title('pie chart example')

```

# Seaborn


```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
plt.figure(figsize = (15, 8))
sns.scatterplot(x = "petal_length", y = "petal_width", hue = 'species', data = df, palette = 'Set2', s = 200)
plt.title('scatter plot: petal length vs petal width')
plt.show()
                
```

```python
plt.figure(figsize = (15, 8))
sns.boxplot(x = 'species', y = 'sepal_length', data = df, hue = 'species', palette = 'Set3')
plt.title('Box plot: Sepal_length by species')
plt.show()
```

```python
plt.figure(figsize = (15, 8))
sns.violinplot(x = 'species', y = 'petal_length',hue = 'species', data = df, palette = 'Pastel1')
plt.title('Violin Plot: Petal length by species')
plt.show()
```

```python
sns.pairplot(df, hue= 'species', palette = 'Set1')
plt.suptitle("pair plot: iris dataset", y = 1.02)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize = (18, 4))
sns.scatterplot(x = "petal_length", y = "petal_width", hue = 'species', data = df, palette = 'Set2', s = 40, ax = axes[0])
axes[0].set_title('scatter plot: petal length vs petal width')

sns.boxplot(x = 'species', y = 'sepal_length', data = df, hue = 'species', palette = 'Set3', ax = axes[1], legend = True)
axes[1].set_title('Box plot: Sepal_length by species')

sns.violinplot(x = 'species', y = 'petal_length',hue = 'species', data = df, palette = 'Pastel1', ax = axes[2], legend = True)
axes[2].set_title('Violin Plot: Petal length by species')
```
