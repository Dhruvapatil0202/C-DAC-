```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

days = [day for day in range(1,21)]
temp = np.random.randint(15,30,20)
humidity = np.random.randint(40,90,20)

```

```python
df = pd.DataFrame({"Days": days, "Temperature": temp, "Humidity": humidity}, dtype="int")

df
```

```python
plt.figure(figsize = (15,8))
sns.lineplot(x="Days",y="Temperature", data=df)
plt.title("Temperature over Days")
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.scatterplot(x="Days",y="Temperature", data=df, hue="Temperature", palette="Set1")
sns.scatterplot(x="Days",y="Humidity", data=df, hue="Humidity", palette="Set2")
plt.title("Temperature vs Humidity over Days")
plt.show()
```

```python
plt.figure(figsize = (8,8))
sns.boxplot(x="Days",y="Temperature", data=df, hue="Days")
plt.title("Temperature over Days")
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.histplot(x="Days", y="Humidity", data=df, hue="Humidity")
plt.title("Humidity Distribution")
plt.show()
```
