# Visualizing Titanic Dataset with Seaborn

### Import Titanic dataset from Seaborn’s built-in datasets. Inspect data by displaying first few rows.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = sns.load_dataset('titanic')
titanic_data
```

```python
titanic_data.head(5)
```

```python
titanic_data.info()
```

```python
titanic_data.isna().sum()
```

#### Removing columns deck, embarked and embark_town
#### Removing rows with np.nan value in column age

```python
titanic_data = titanic_data.drop(columns = ['deck', 'embark_town', 'embarked'], axis = 1)
titanic_data.isna().sum()
```

```python
titanic_data = titanic_data.dropna(subset = ['age']).reset_index(drop = True)
titanic_data.isna().sum()
```

```python
titanic_data.info()
```

## Create a Count Plot: 
### Use sns.countplot to create a count plot showing number of passengers who survived versus  thos  who did not. Add appropriate labels and a title to the plot.


```python
plt.figure(figsize = (15, 8))
sns.countplot(x = 'survived', data = titanic_data, hue = "pclass", palette = 'Set1')
plt.title("Count of survived passengers vs succumed passengers")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.xticks(ticks = [0, 1], labels = ["Did not Survive", "Survived"])
plt.legend(title = 'Passenger Class', labels = ['First Class', 'Second Class', 'Thrid Class'])
plt.show()
```

## Create a Box Plot:




### Use sns.boxplot to create a box plot of passenger ages grouped by survival status (Survived or Not Survived). Ensure that each survival status is represented by a different color in the box plot.

```python
plt.figure(figsize = (15, 8))
sns.boxplot(x = 'survived', y = 'age', data= titanic_data, hue = "sex", palette = 'Set2')
plt.xticks(ticks = [0, 1], labels = ['Not Survived', 'Survived'])
plt.xlabel("Survival Status")
plt.ylabel("Age")
plt.title("Box Plot of passenger Ages by Survival Status")
plt.show()
```

## Create a Violin Plot:


### Use sns.violinplot to create a violin plot of the fares paid, grouped by passenger class (Pclass). Add appropriate labels and a title to the plot.

```python
plt.figure(figsize = (15, 8))
sns.violinplot(x = 'pclass', y = 'fare', hue = 'survived', palette = 'Set1', data = titanic_data)
plt.xlabel("Class of Passenger")
plt.xticks(ticks = [0, 1, 2], labels = ['First Class', 'Second Class', 'Thrid Class'])
plt.ylabel("Fair paid by Passenger")
plt.show()
```

## Create a Histogram: 

### Use sns.histplot to create a histogram of the ages of the passengers. Add a density plot to the histogram to visualize the distribution of ages.

```python
plt.figure(figsize = (15, 8))
sns.histplot(data= titanic_data, x ='age', kde = True, bins = 30, color= 'blue')
plt.title("Density plot of Passenger Ages")
plt.show()
```

### Bring all graphs together to display in sections on 1 screen.

```python
fig, axis = plt.subplots(2, 2, figsize = (15, 8))

# Count plot
sns.countplot(x = 'survived', data = titanic_data, hue = "pclass", palette = 'Set1', ax = axis[0, 0])
axis[0, 0].legend(title = 'Passenger Class', labels = ['First Class', 'Second Class', 'Thrid Class'])

# Box Plot
sns.boxplot(x = 'survived', y = 'age', data= titanic_data, hue = "sex", palette = 'Set2', ax = axis[0, 1])

# Violin Plot
sns.violinplot(x = 'pclass', y = 'fare', hue = 'survived', palette = 'Set1', data = titanic_data, ax = axis[1, 0])

# Histogram
sns.histplot(data= titanic_data, x ='age', kde = True, bins = 30, color= 'blue', ax = axis[1, 1])


plt.show()
```
