```python
import pandas as pd
import random
```

```python

Name = ['name' + str(i) for i in range(1, 11)]
Id = [12400200 + i for i in range(1, 11)]
Company = [chr(i) for i in range(ord('a'), ord('a') + 10)]
Salary = [10000 + 1000 * i for i in range(1, 11)]
```

```python
df1 = pd.DataFrame({'name' : Name, 'id': Id})
df1
```

```python
df2 = pd.DataFrame({'company' : Company, 'salary': Salary})
df2
```

```python
new_df = pd.concat([df1, df2], axis = 1)
new_df
```
