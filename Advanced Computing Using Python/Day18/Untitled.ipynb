```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = r"C:\Users\DAI.STUDENTSDC\Desktop\DAI 2024\Day18\dataset\apple_stock_data.csv"
```

```python
class StockData:
    def __init__(self,file_path):
        self.file_path = file_path
        self.data = pd.DataFrame()

    def load_data(self):
        try:
            self.data = pd.read_csv(self.file_path)
        except:
            raise FileNotFoundError("File does not exists!")


    def sort_data(self):
        sorted_data = self.data.sort_values(by="Date", inplace=False)
        return sorted_data

    def visualize_data(self):
        sorteddata = self.sort_data()
        plt.figure(figsize=(16,8))
        plt.scatter(x="Date", y="Close", data=sorteddata, color="blue")
        plt.title("Stock Data")
        plt.xlabel("Date")
        plt.ylabel("Close")
        plt.show()

stockmarket = StockData(filepath)

stockmarket.load_data()

stockmarket.visualize_data()
        
```
