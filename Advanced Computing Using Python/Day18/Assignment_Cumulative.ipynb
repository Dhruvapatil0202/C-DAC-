```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = r"C:\Users\DAI.STUDENTSDC\Desktop\DAI 2024\Day18\dataset\apple_stock_data.csv"
```

```python
class StockData:
    # def __init__(self, filepath, date, open, high, low, colse, volume):
    def __init__(self, filepath):
        
        self.df = pd.DataFrame()
        self.filepath = filepath
        
    def load_data(self):
        try:
            self.df = pd.read_csv(self.filepath)
        except: 
            raise FileNotFoundError("The file was not there")
            
        return self.df

    def sort_data(self, column_name, inplace = False):
        if column_name not in self.df.columns: 
            raise ValueError(f"The column {column_name} does not exist in the dataframe")
        new_df = self.df.sort_values(by = [column_name])
        if inplace:
            self.df = new_df
        return new_df
    
    def visualize_data(self):

        plt.figure(figsize = (15, 8))
        plt.scatter(x = "Date", y = "Close", data = self.df, color = 'black')
        plt.title("Closeing Price over time")
        plt.xlabel("Date")
        plt.ylabel("Closeing price")
        plt.show()

    def visualize_short(self):

        temp_df = self.df.iloc[::100, ::]
        plt.figure(figsize = (15, 8))
        plt.scatter(x = "Date", y = "Close", data = temp_df, color = 'black')
        plt.title("Closeing Price over time")
        plt.xlabel("Date")
        plt.ylabel("Closeing price")
        plt.show()
        
```
