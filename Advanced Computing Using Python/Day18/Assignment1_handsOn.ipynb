```python
import pandas as pd
```

```python
students = pd.read_csv("students_df.csv")
students
```

```python
extra_scores = pd.read_csv("extra_scores_df.csv")
extra_scores.head()
```

```python
students = students.dropna(subset=["Age"]).reset_index()
students
```

```python
extra_scores["Additional_Score"] = extra_scores["Additional_Score"].fillna(extra_scores["Additional_Score"].mean())
extra_scores["Additional_Score"] = extra_scores["Additional_Score"].apply(lambda x : float(int(x)))
extra_scores
```

## Filter the students_dataframe 

students scored more than 80 and are older than 27 years

```python
students
```

```python
more_than_70 = students["Score"] >= 80
older_than_20 = students["Age"] > 27
filtered_data = students[more_than_70 & older_than_20]
filtered_data.reset_index()
```

Adding new column 

```python
new_dataframe = students.copy()
new_dataframe["Final Score"] = new_dataframe["Score"] + 10
new_dataframe
```

Remove the gender column from dataframe


```python
new_dataframe1 = students.copy()
new_dataframe1 = new_dataframe.drop(columns = ["Gender"])
new_dataframe1
```

sort dataframe in decreasing order of score 

```python
new_df2 = new_dataframe.copy()
new_df2 = new_df2.sort_values(by = ["Final Score"], ascending = False).drop(columns = 'index', axis = 0)
new_df2
```

Mergeing the Additional_scores df on students df on ID such that data in students df is not changed in any way

```python
new_df3 = pd.merge(left = students,
                  right = extra_scores,
                  how = 'left',
                  left_on = 'ID',
                  right_on = 'ID')
new_df3
```

```python
new_df3["Total Score"] = new_df3["Score"] + new_df3["Additional_Score"]
new_df3
```
