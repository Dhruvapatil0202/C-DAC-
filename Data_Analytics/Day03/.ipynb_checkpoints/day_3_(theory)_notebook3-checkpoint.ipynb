```python

# Data Visualization is very important to quickly understand trends and relationships in the data

# Matplotlib : One of the most popular plotting libraries in Python
# -> Grandfather of plotting and visualization libraries in python
# -> Seaborn/Pandas built-in visualization are buildt on top of Matplotlib
# -> heavily inspired by the plotting functions in the MatLab progrmming language 
# -> Almost any type of plot can be created and customized
# -> Can be confusing at first: Two approaches: 1. Functional 2. OOP
# -> Matplot basec plots: Functional, Figures and subplots: OOP
# -> Main Goals 
# 1. plot a functional relationship, eg. y = 2x
# 2. plot a relationship between raw data points: x = [1, 2, 3, 4] and y = [2, 4, 6, 8]
```

```python


import numpy as np
import matplotlib.pyplot as plt
```

```python
# 
x = np.arange(1, 10)
y = np.arange(1, 19)

a, b = np.random.randint(5, 15, 10), np.random.randint(50, 100, 10)
# Select the first 9 elements from y to match x-axis length
y = y[:9] # Slicing to get elements from index 0 (inclusive) to 9 (exclusive)

plt.plot(x, y)
plt.xlabel('x axis title here')
plt.ylabel('y axis title here')
plt.title('Plot title Here')
plt.scatter(x, y, s = 50)
plt.show()
```

```python

# Exporting a plot
plt.plot(x, y)
plt.savefig("example.png")
```
