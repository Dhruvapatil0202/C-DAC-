```python


# Unique 
print(tips['size'].unique())
print(tips['size'].nunique())
print(tips['time'].unique())
```

```python


# Between: options for inclusive are both, left, right, neither
print(tips['total_bill'].between(10, 20, inclusive = "both")) # True vallse values
print(tips[tips['total_bill'].between(10, 20, inclusive = "both")])  # Actual Data 
```

```python


# Sample data, randomly selected
print(tips.sample(5))   # 5 rows
print("\n" * 4)
print(tips.sample(frac = 0.05))   # 5% of rows
```

```python


# nlargest, nsmallest
print(tips.nlargest(5, 'tip'))
print("\n" * 4)
print(tips.nsmallest(5, 'tip'))
```

```python


# GroupBy functionality
# 1. Splitting: Data is first split into groups
# 2. Applying: A function is applied to each group independently
# 3. Combining: The results of the functio applications are then combined into a new DataFrame
```

```python
# 1. Splitting 
tips_grouped = tips.groupby('gender')
tips_grouped
```

```python
# 2. Applying
# Calculate the mean tip for each gender
mean_tip_by_gender = tips_grouped['tip'].mean()
mean_tip_by_gender
```

```python
print(tips_grouped['tip'].min())
print("\n")
print(tips_grouped['tip'].max())
print("\n")
print(tips_grouped['tip'].std())
print("\n")
print(tips_grouped['tip'].var())
print("\n")
print(tips_grouped['tip'].count())
print("\n")
print(tips_grouped['tip'].sum())
print("\n")
# print(tips_grouped['tip'].min())
```

```python


# Creating groups on multiple columns
tips_grouped_2 = tips.groupby(['gender', 'day'])
print(tips_grouped_2['tip'].mean())
```

```python


# agg function
# Appply multiple aggregation functions to different columns when grouping data
# Calculate both the mean and sum of the 'tip' column for each combination of 'gender' and 'day'

# Group the DataFrame by both 'gender' and 'day' and calculate the mean and sum of 'tip' or each combination

tips_grouped = tips.groupby(['gender', 'tip']).agg({'tip' : ['mean', 'sum']}).reset_index()
print(tips_grouped)
```

```python
tips_grouped.columns = ['Gender', 'Tip', 'Avg Tip', 'Total Tip']
tips_grouped
```
