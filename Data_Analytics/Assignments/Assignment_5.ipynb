```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "Programming languages": ["Java", "Python", "PHP", "JavaScript", "C#", "C++"],
    "Popularity": [22.2, 17.6, 8.8, 8, 7.7, 6.7]
})
df.plot(kind = 'bar')
plt.show()
print("\n\n\n\n")
```

```python
df = pd.DataFrame({
    "Programming languages": ["Java", "Python", "PHP", "JavaScript", "C#", "C++"],
    "Popularity": [22.2, 17.6, 8.8, 8, 7.7, 6.7]
})

plt.barh(df["Programming languages"], df["Popularity"], color = "green", edgecolor = 'black')
plt.xlabel("Populaarity")
plt.ylabel("Programming Languages")
plt.show()

print("\n\n\n\n")
```

```python
df = pd.DataFrame({
    "Programming languages": ["Java", "Python", "PHP", "JavaScript", "C#", "C++"],
    "Popularity": [22, 17.6, 8.8, 8, 7.7, 6.7]
})

plt.bar(df["Programming languages"], df["Popularity"], data=df, edgecolor="black", align="edge")
plt.grid(which="major", color="red", linestyle="-")
plt.minorticks_on()
plt.grid(which="minor", color="grey", linestyle=":", linewidth=0.4)
plt.gca().xaxis.set_ticks_position('top')  # Set ticks on the top
plt.gca().xaxis.set_label_position('top')  
plt.gca().invert_yaxis()

plt.xlim("Java","")

plt.xlabel("Programming languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Language \nWorldwide, Oct 2017 compared to a year ago")

```

```python
df = pd.DataFrame({
    "Programming languages": ["Java", "Python", "PHP", "JavaScript", "C#", "C++"],
    "Popularity": [22, 17.6, 8.8, 8, 7.7, 6.7]
})

plt.grid(which="major", color="red", linestyle="-", alpha=0.2)
plt.minorticks_on()
plt.grid(which="minor", color="lightgrey", linestyle=":", linewidth=0.4, alpha=0.7)

bars = plt.bar(df["Programming languages"], df["Popularity"], color="blue", data=df, edgecolor="black", width=0.8, zorder=5, align="edge")


plt.bar_label(bars, fmt='%.6f', label_type='edge', padding=3)
plt.ylim(0,25)
plt.xlim("Java", "")

plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Language \nWorldwide, Oct 2017 compared to a year ago")
plt.show()
```
