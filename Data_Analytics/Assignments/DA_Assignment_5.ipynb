```python
import numpy as np
import matplotlib.pyplot as plt
num_set = [{'Language': 75, 'Science': 88, 'Math': 96},
           {'Language': 71, 'Science': 95, 'Math': 92},
           {'Language': 75, 'Science': 90, 'Math': 89}]

lan_guage = [['Language', 'Science', 'Math'], 
             ['Science', 'Math', 'Language'], 
             ['Math', 'Language', 'Science']]

colors = ["r", "g", "b"]
names = sorted(num_set[0].keys())

values = np.array([[data[name] for name in order] for data, order in zip(num_set, lan_guage)])
lefts = np.insert(np.cumsum(values, axis=1), 0, 0, axis=1)[:, :-1]
orders = np.array(lan_guage)
bottoms = np.arange(len(lan_guage))

for name, color in zip(names, colors):
    idx = np.where(orders == name)
    value = values[idx]
    left = lefts[idx]
    plt.barh(bottoms, width=value, left=left, height=0.8, color=color, label=name, edgecolor = 'black')

plt.yticks(bottoms, ["Student-%d" % (t + 1) for t in bottoms])
plt.legend(loc="best", bbox_to_anchor=(1.0, 1.00))

plt.subplots_adjust(right=0.75)

plt.grid(which="major", color="grey", linestyle="-", alpha=0.2)
plt.minorticks_on()
plt.grid(which="minor", color="lightgrey", linestyle=":", linewidth=0.4, alpha=0.7)

plt.xlabel('Scores')
plt.title('Scores by Subject for Each Student')
plt.show()
```

```python
data = [[ 3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039,
7.1123587, 12.77792868, 3.44773477],
[ 11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195,
6.79685302, 7.24578743, 3.69371847],
[ 3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182,
4.63832778, 11.16849999, 8.56883433],
[ 4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185,
10.97567589, 3.98287652, 8.80552122]]

df = pd.DataFrame(data)
df = df.T
df.index = ["G1", "G2", "G3", "G4", "G5", "G6", "G7", "G8"]

ser1, ser2, ser3, ser4 = df.iloc[:, 0], df.iloc[:, 1], df.iloc[:, 2], df.iloc[:, 3]

plt.barh(y = df.index, width = ser1, color = 'red', edgecolor = 'black')
plt.barh(y = df.index, width = ser2, color = 'green', edgecolor = 'black', left = ser1)
plt.barh(y = df.index, width = ser3, color = 'white', edgecolor = 'black', left = ser1 + ser2)
plt.barh(y = df.index, width = ser4, color = 'purple', edgecolor = 'black', left = ser1 + ser2 + ser3)

total_widths = ser1 + ser2 + ser3 + ser4

for i in range(len(df)):

    positions = [0, ser1.iloc[i], ser1.iloc[i] + ser2.iloc[i], ser1.iloc[i] + ser2.iloc[i] + ser3.iloc[i]]
    widths = [ser1.iloc[i], ser2.iloc[i], ser3.iloc[i], ser4.iloc[i]]
    
    for j in range(4):
        percentage = (df.iloc[i, j] / df.iloc[:, j].sum()) * 100 if total_widths.iloc[i] > 0 else 0
        if widths[j] > 0:  
            plt.text(positions[j] + widths[j] / 2, i, f'{percentage:.1f}%', 
                     ha='center', va='center', color='black' if j < 3 else 'yellow')
plt.show()
```

```python
hatchs = ['/', 'x', '|', 'o', 'O', '.', '*', "+", "\\", "-"]
xind = list(range(10))
value = 3.0

fig, ax = plt.subplots(figsize = (15, 8))

bars = ax.bar(xind, height = [value] * len(xind), edgecolor = 'black', color = "lightblue", width = 0.95)
for bar_ind in xind:
    bars[bar_ind].set_hatch(hatchs[bar_ind])

plt.ylabel('Values')
plt.title('Bar Chart with Modified First Bar')
plt.show()
```
