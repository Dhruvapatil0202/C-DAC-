```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {
    'Date': ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05', '2024-10-06', '2024-10-07'],
    'Open': [1500.0, 1505.0, 1515.0, 1525.0, 1530.0, 1540.0, 1545.0],
    'High': [1510.0, 1520.0, 1530.0, 1540.0, 1550.0, 1555.0, 1560.0],
    'Low': [1495.0, 1500.0, 1510.0, 1520.0, 1525.0, 1535.0, 1540.0],
    'Close': [1505.0, 1515.0, 1525.0, 1530.0, 1540.0, 1545.0, 1550.0],
    'Volume': [1200000, 1300000, 1400000, 1100000, 1500000, 1250000, 1150000]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
main_df = df.copy()
print(df)
```

1. Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.

```python
start_date, end_date = "2024-10-01", "	2024-10-04"
df = main_df.loc[(main_df['Date'] >= start_date) & (main_df['Date'] <= end_date)]

plt.plot(df["Date"], df["Open"], label = "Open", color = 'green')
plt.plot(df["Date"], df["High"], label = "High", color = 'red')
plt.plot(df["Date"], df["Low"], label = "Low", color = 'blue')
plt.plot(df["Date"], df["Close"], label = "Close", color = 'orange')

ax = plt.gca()

ax.set_xticks(df["Date"])
plt.xticks(rotation = 45)

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Alphabet Inc. between two specific dates")
plt.legend()
plt.show()
print("\n\n\n\n")
```

13. Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of Alphabet Inc. between two specific dates.

```python
date_range = pd.date_range(start='2024-10-01', periods=30, freq='D')

data = {
    'Date': date_range,
    'Open': np.random.uniform(1500, 1600, 30),
    'High': np.random.uniform(1510, 1610, 30),
    'Low': np.random.uniform(1490, 1590, 30),
    'Close': np.random.uniform(1500, 1600, 30),
    'Volume': np.random.randint(1000000, 1600000, 30),
    'Adjusted Close': np.random.uniform(1500, 1600, 30)
}

mdf = pd.DataFrame(data)
mdf["Date"] = pd.to_datetime(mdf["Date"])

```

```python
start_date, end_date = "2024-10-02", "2024-10-18"
df = mdf.loc[(mdf['Date'] >= start_date) & (mdf['Date'] <= end_date)]

df.set_index("Date", inplace = True)

fig, axis = plt.subplots(3, 2, figsize = (15, 23))
open_fig = axis[0][0]
close_fig = axis[0][1]
high_fig = axis[1][0]
low_fig = axis[1][1]
adj_fig = axis[2][0]
vol_fig = axis[2][1]

open_fig.plot(df["Open"], label = "Open", color = 'green', lw = 1.5)
open_fig.set_xlabel("Date")
open_fig.set_ylabel("Open")
open_fig.set_title("open price plot")
open_fig.legend()

close_fig.plot(df["Close"], label = "Close", color = 'red', lw = 1.5)
close_fig.set_xlabel("Date")
close_fig.set_ylabel("Close")
close_fig.set_title("Close price plot")
close_fig.legend()

high_fig.plot(df["High"], label = "High", color = 'blue', lw = 1.5)
high_fig.set_xlabel("Date")
high_fig.set_ylabel("High")
high_fig.set_title("High price plot")
high_fig.legend()

low_fig.plot(df["Low"], label = "Low", color = 'orange', lw = 1.5)
low_fig.set_xlabel("Date")
low_fig.set_ylabel("Low")
low_fig.set_title("Low price plot")
low_fig.legend()

adj_fig.plot(df["Adjusted Close"], label = "Adjusted Close", color = 'Purple', lw = 1.5)
adj_fig.set_xlabel("Date")
adj_fig.set_ylabel("Adjusted Close")
adj_fig.set_title("Adjusted Close price plot")
adj_fig.legend()

vol_fig.plot(df["Volume"], label = "Volume", color = 'black', lw = 1.5)
vol_fig.set_xlabel("Date")
vol_fig.set_ylabel("Volume")
vol_fig.set_title("volume plot")
vol_fig.legend()

plt.show()
print("\n\n\n\n")
```

14. Write a Pandas program to create a plot of adjusted closing prices, thirty days and forty days simple moving average of Alphabet Inc. between two specific dates.


```python
start_date, end_date = "2024-10-02", "2024-10-25"
df = mdf.loc[(mdf['Date'] >= start_date) & (mdf['Date'] <= end_date)]

# Insted of 30 and 40 the window of 3 and 4 is taken for the dataset is small

ser3 = df.rolling(window = 3)['Adjusted Close'].mean().dropna()
ser4 = df.rolling(window = 4)['Adjusted Close'].mean().dropna()

fig, axes = plt.subplots(2, 1, figsize = (15, 16))
fig1 = axes[0]
fig2 = axes[1]

fig1.plot(ser3, label = "3 days moving average", color = 'red', lw = 2)
fig1.set_ylabel("Avg. Adjusted Close")
fig1.set_title("3 days moving average")

fig2.plot(ser4, label = "4 days moving average", color = 'blue', lw = 2)
fig2.set_ylabel("Avg. Adjusted Close")
fig2.set_title("4 days moving average")

plt.show()
```


15. Write a Pandas program to create a plot of adjusted closing prices, 30 days
simple moving average and exponential moving average of Alphabet Inc.
between two specific dates.


```python
start_date, end_date = "2024-10-02", "2024-10-25"
df = mdf.loc[(mdf['Date'] >= start_date) & (mdf['Date'] <= end_date)]

# Insted of 30 days window, using window = 3 for limited dataset

ser3 = df.rolling(window = 3)['Adjusted Close'].mean().dropna()
ser4 = df['Adjusted Close'].ewm(span=3, adjust=False).mean().dropna()

fig, axes = plt.subplots(2, 1, figsize = (15, 16))
fig1 = axes[0]
fig2 = axes[1]

fig1.plot(ser3, label = "3 days moving average", color = 'red', lw = 2)
fig1.set_ylabel("Avg. Adjusted Close")
fig1.set_title("3 days moving average")

fig2.plot(ser4, label = "Exponential Average", color = 'blue', lw = 2)
fig2.set_ylabel("Avg. Adjusted Close")
fig2.set_title("Exponential Average")

plt.show()
```


17. Write a Pandas program to create a plot to visualize daily percentage returns of Alphabet Inc. stock price between two specific dates.

```python
start_date, end_date = "2024-10-05", "2024-10-25"
df = mdf.loc[(mdf['Date'] >= start_date) & (mdf['Date'] <= end_date)]

ser = ((df.Close - df.Open) / df.Open) * 100

fig, axis = plt.subplots(2, 1, figsize = (15, 16))
fig1 , fig2 = axis[0], axis[1]

fig1.bar(df['Date'], ser)
fig1.set_xlabel("Date")
fig1.set_ylabel("Percentage")
fig1.set_title("Percent Gain")

fig2.plot(df['Date'], ser, label='Percent Gain', color='blue', lw=2)
fig2.set_xlabel("Date")
fig2.set_ylabel("Percentage gain")
fig2.set_title("Percent Gain")
fig2.legend()

plt.show()
```


18. Write a Pandas program to plot the volatility over a period of time of Alphabet
Inc. stock price between two specific dates.

```python
start_date, end_date = "2024-10-05", "2024-10-25"
df = mdf.loc[(mdf['Date'] >= start_date) & (mdf['Date'] <= end_date)].copy()

df['Date'] = pd.to_datetime(df['Date'])

df['Return'] = df['Close'].pct_change()

window_size = 7
df['Volatility'] = df['Return'].rolling(window=window_size).std()

plt.figure(figsize=(15, 8))

plt.plot(df['Date'], df['Volatility'], label=f'Volatility ({window_size}-day rolling SD)', color='blue', lw=2)

plt.xlabel("Date")
plt.ylabel("Volatility (Standard Deviation of Returns)")
plt.title(f"Stock Volatility Over Time (Window: {window_size} Days)")
plt.axhline(0, color='red', linestyle='--', linewidth=0.8)
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
