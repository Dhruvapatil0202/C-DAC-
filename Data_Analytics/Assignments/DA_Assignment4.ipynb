```python
df = pd.DataFrame(data = {
    "x-axis" : np.arange(10, 31, 10),
    "line_1" : np.array([40, 10, 30]),
    "line_2" : np.array([20, 40, 10])
})

plt.plot(df["x-axis"], df["line_1"], linewidth= 3, color = 'blue', label = 'line-1')
plt.plot(df["x-axis"], df["line_2"], linewidth= 5, color = 'red', label = 'line-2')
plt.ylabel("y-axis")
plt.legend()
plt.grid()
plt.show()
print("\n\n\n\n")
```

```python
df = pd.DataFrame(data = {
    "x-axis" : np.arange(10, 31, 10),
    "line_1" : np.array([40, 10, 30]),
    "line_2" : np.array([20, 40, 10])
})

plt.plot(df["x-axis"], df["line_1"], linewidth= 3, color = 'blue', label = 'line-1', linestyle = ":")
plt.plot(df["x-axis"], df["line_2"], linewidth= 5, color = 'red', label = 'line-2', linestyle = "--")
plt.ylabel("y-axis")
plt.legend()
plt.grid()
plt.show()
print("\n\n\n\n")
```

```python
df = pd.DataFrame({
    'x-axis': np.array([2, 3, 4, 5, 6, 7, 8, 9]),
    'y-one' : np.array([1, 5, np.nan, 10, 17, np.nan, 20, np.nan]),
    'y-two' : np.array([np.nan, 2, 6, np.nan, 13, 20, np.nan, 24]),
})
plt.axis([0, 10, 0, 30])
plt.scatter(df['x-axis'], df['y-one'], marker = "*")
plt.scatter(df['x-axis'], df['y-two'], color = 'red')
plt.show()
```

```python
ser0 = np.arange(0, 5, 0.2)
ser1 = ser0 * 1.2
ser2 = ser0 ** 2
ser3 = ser0 ** 3

plt.axis([0, 5, 0, 120])
plt.scatter(ser0, ser1, color = 'green', marker = '_')
plt.scatter(ser0, ser2, color = 'blue', marker = 's')
plt.scatter(ser0, ser3, color = 'red', marker = '^')
plt.show()
```
