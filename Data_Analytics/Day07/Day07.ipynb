Variance - how far values are from the mean
high variance- data is more scattered
low variance - data is less scattered

Calculate variance
1. calculate mean
2. subtract the mean from each value and wquare the resullt
3. the avg of these squared values is the variance

Population variance -  variance of Complete data
sample variance - variance of sample data taken from complete data.
Usually we work on sample data cause it is nearly impossible to get the Complete data

Remember for accuracy point of view take n-1 at denominator to calculate sample variance

We square the difference while calculating the variance because positive and negative values will cancel each other and thus it will affect the result resulting in the wrong variance. and other reason is the values which are more away from the mean will result in getting higher variance values. So this helps in getting approximate idea about the how the data is scattered.

Problem with variance is interpretation  becomes difficult

Variance : By squaring the differences:
1. +ve and -ve effects donot cancel each other
2.    higher variances are highlighted even further
But..
1.   Scale of variance is the square of the orignal data scale

To bring the variance to its orignal scale we take the square root and we call the resultant quantity as "Standard Deviation"

Quantifying or measuring variance is tough

Scale of standard deviation is same as the orignal data.

eg- mean = 69, sd = 11
mean+-sd, mean+-2sd, mean+-3sd       will cover almost 100% of values


We can fimd how much standard deviation away a point is away from the mean
formulae : = (Our data point - Mean)/ standard deviation
positive and negative signs matter cause it decides on which side of the mean the deviation is- left or right

Coefficient of Variation (CV por COV)-  Size of standard variation deviation related to mean
CV : standardized and unitless(relative)
sd : expressed in the orignal scale/ units of data(absolute)

High CV -> Large SD compared to mean

CV = (sd/mean)*100

It is used to compare with CV's of other groups, then only it will make sense.

High consistency means low CV
Low consistency means high CV

*Skewness 
represents asymmetry of a distribution around its mean
No skew - data in the form of bell shape is called Zero skew
Right skew - most data equal to low marks                                         Interpretation - most values are low
mean>median
mean skewed to the right
outliers to right
tail comes later

Left skew -  most students equal to high marks                                     Interpretation - most values are high
mean< median
mean skewed to the left
outliers to left
tail comes first

skew= where is the skew(drop curve) , not where is the data



```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Generate random data with a larger sampe sizwe
sample_size =  100000
ratings = np.random.randint(1,6,sample_size)
```

```python
# Modify data for right skew
right_skew_data = np.concatenate((ratings, np.random.randint(1,2, sample_size//2)))
```

```python
# Modify data for left skew
left_skew_data = np.concatenate((ratings, np.random.randint(4,6,sample_size //2)))
```

```python
# Function to calculate mean, median, and skewness
def  calculate_stats(data):
    mean = np.mean(data)
    median = np.median(data)
    skewness = pd.Series(data).skew()
    return mean, median, skewness
```

```python
# Function to plot histogram with mean and median lines
def plot_histogram(ax,data,title):
    ax.hist(data, bins=np.arange(1,7) - 0.5, rwidth= 0.8, alpha=0.7)
    ax.set_title(title)
    ax.set_xlabel('Rating')
    ax.set_ylabel('Frequency')
    ax.set_xticks(range(1,6))
    ax.grid(axis='y', alpha=0.5)

    # Calculate mean , median, and skewness
    mean, median, skewness = calculate_stats(data)
    ax.axvline(x=mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
    ax.axvline(x= median, color='blue', linestyle='--', label=f'Median: {median}')
    ax.legend()

    # Print mean, median, skewness
    print(f"{title}:")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median}")
    print(f"Skewness: {skewness:.2f}")
    print()
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,5))

plot_histogram(axs[0], ratings, 'No Skew')
plot_histogram(axs[1], right_skew_data, 'Right Skew')
plot_histogram(axs[2], left_skew_data, 'Left Skew')
plt.tight_layout()
plt.show()
```

Skewness Formulae
--------------------------------------------
1. Pearson's First Coefficient of Skewness(Mode Skewness): (Mean - Median) / Standard Deviation
2. Another Variation: 3 (Mean - Median) / Standard Deviation
3. Fisher-Pearson Coefficient of Skewness (Moment Skewness): ()
4. Quantile-Based Skewness: (Q3 + Q1 - 2 * Median) / (Q3 - Q1)

if value of skewness is negetive, it is right skew. 
if value of skewness is positive, it is left skew.
if value of skewness is near to zero it is now skewed.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
airline_df  = pd.read_csv("airline_dec_2008_50k.csv")
distance = airline_df["Distance"]

mean_distance = distance.mean()
median_distance = distance.median()
skewness_distance = distance.skew()

plt.figure(figsize = (8, 6))
sns.histplot(distance, kde = True, color = 'skyblue', bins = 30)
plt.axvline(mean_distance, color = 'red', linestyle = "--", label = f"Mean: {mean_distance:.2f}")
plt.axvline(median_distance, color = 'blue', linestyle = "--", label = f"Median : {median_distance:.2f}")
plt.title(f"Distribution of Distance\nSkewness: {skewness_distance:.2f}")
plt.xlabel("Distance")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```
