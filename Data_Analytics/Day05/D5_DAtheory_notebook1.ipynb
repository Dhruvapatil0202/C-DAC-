```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a = np.linspace(1,10, 11)
b = a ** 4

x = np.arange(0,10)
y = 2 * x 
```

```python
fig = plt.figure()
axes=  fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height
axes.plot(x, y)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b)
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (15, 8))
axes[0][0].plot(a, b)
axes[1][0].plot(x, y)
axes[0][1].plot(b, a)
axes[1][1].plot(y, x)

fig.subplots_adjust(left = None,
                    bottom = None,
                    right = None,
                    top = None,
                    wspace = 0.9,
                    hspace = 0.1,)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize = (8, 6))
plt.scatter(x, y, color = 'blue', alpha = 0.6)
plt.title("Scatter Plot using matplotlib", fontsize = 15)
plt.xlabel("x-axis", fontsize = 12)
plt.ylabel("y-axis", fontsize = 12)
plt.grid(True)
plt.show()
```

```python
# Creation of scatterplot using seaborn
sns.set(style = "whitegrid") # set style with white grid background
plt.figure(figsize = (8, 6))
sns.scatterplot(x = x, y = y)
plt.title("Scatter Plot using Seaborn", fontsize = 15)
plt.show()
```

```python
df = pd.read_csv("heart.csv")
ages = df.groupby("age").median().reset_index()
# 1. Group on the age column 2. Calculate median for each numeric column on the age group 3. Reset index to reset to numeric index
print(df.head())
print(df.info())
```

```python
# Simple Pandas
ages.plot.line("age", "chol")
plt.show()
```

```python
#matplotlib
#
fig, ax = plt.subplots()
ax.plot(ages["age"],ages["chol"], ls=":", lw=1.7)
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
plt.show()
```

```python
#seaborn
# sns.lineplot(x=ages age"], y=ages["chol"], data = "ages", linestyle=":", linewidth=1.7)
sns.lineplot(x = "age", y = "chol", data = ages, linestyle = ":", linewidth = 1.7)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

```python
df = pd.read_csv("dm_office_sales.csv")

## Countplot: A simple plot, it merely shows the total count of rows per category
# A histogram across a categorical, instead of quantativem vvariable
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(x='division', data = df)
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(x='level of education', data=df)
plt.show()
```

```python
# Breakdown within another category with hue
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(x='level of education', data=df, hue='training level')
plt.show()
```

```python
# Using palette
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(x='level of education', data=df, hue='training level', palette='Set1')
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=200)
#Paired would be a good choice if there was a distinct jump from 0 and 1 to 2 and 3
sns.countplot(x='level of education', data= df, hue='training level', palette="Paired")
plt.show()
```
