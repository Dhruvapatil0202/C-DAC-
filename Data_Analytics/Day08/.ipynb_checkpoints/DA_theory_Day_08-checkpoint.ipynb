```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("weight-height.csv")
```

```python
# Seperate data based on gender
df_male = df[df['Gender'] == "Male"]
df_female = df[df['Gender'] == 'Female']
```

```python
# Calculate correlation 
corr_overall = df[['Height', "Weight"]].corr()
corr_male = df_male[["Height", "Weight"]].corr()
corr_female = df_female[["Height", "Weight"]].corr()
```

```python
# Calculat ecovariance
cov_overall = df[['Height', 'Weight']].cov()
cov_male = df_male[['Height', 'Weight']].cov()
cov_female = df_female[['Height', 'Weight']].cov()
```

```python
print("Overall Correlation between Height and Weight: ")
print(corr_overall)

print(f"\nMale Correlation between Height and Weight: \n{corr_male}")
print(f"\nFemale Correlation between Height and Weight: \n{corr_female}")
print(f"\nOverall Covariance between Height and Weight: \n{cov_overall}")
print(f"\nMale Covariance between Height and Weight: \n{cov_male}")
print(f"\nFemal Covariance between Height and Weight: \n{cov_female}")
```

```python
fig, axes = plt.subplots(1, 3, figsize = (18, 6))

axes[0].scatter(df["Height"], df["Weight"], color = 'blue', alpha = 0.5)
axes[0].set_title(f"Overall Correlation: {corr_overall.loc["Height", "Weight"]:.2f}")
axes[0].set_xlabel("Height")
axes[0].set_ylabel("Weight")
axes[0].grid(True)

axes[1].scatter(df["Height"], df["Weight"], color = 'green', alpha = 0.5)
axes[1].set_title(f"Male Correlation: {corr_male.loc["Height", "Weight"]:.2f}")
axes[1].set_xlabel("Height")
axes[1].set_ylabel("Weight")
axes[1].grid(True)

axes[2].scatter(df["Height"], df["Weight"], color = 'red', alpha = 0.5)
axes[2].set_title(f"Female Correlation: {corr_female.loc["Height", "Weight"]:.2f}")
axes[2].set_xlabel("Height")
axes[2].set_ylabel("Weight")
axes[2].grid(True)

plt.show()
```
