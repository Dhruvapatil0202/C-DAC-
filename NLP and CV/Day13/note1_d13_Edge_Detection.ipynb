```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img  = cv2.imread('sachin.jpg', 0)
```

```python
help(cv2.Sobel)
```

Define sobel operators

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = 3)
```

```python
sobelx
```

```python
final_sobel = cv2.add(sobelx, sobely)
```

```python
final_sobel
```

```python
thresh, binary_sobel = cv2.threshold(final_sobel, 60, 255, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize =(16,9))
plt.subplot(1,4,1)
plt.title('Orignal Image')
plt.imshow(img, cmap = 'gray')

plt.subplot(1,4,2)
plt.title('Sobel X Image')
plt.imshow(sobelx, cmap = 'gray')

plt.subplot(1,4,3)
plt.title('Sobel-Y Image')
plt.imshow(sobely, cmap = 'gray')

plt.subplot(1,4,4)
plt.title('Edge Detected Image')
plt.imshow(binary_sobel, cmap = 'gray')
```

Canny Edge Detection

```python
help(cv2.Canny)
```

```python
plt.figure(figsize =(16,9))
plt.subplot(1,2,1)
plt.title('Orignal Image')
plt.imshow(img, cmap = 'gray')

plt.subplot(1,2,2)
plt.title('Edge Detected')
plt.imshow(edges, cmap = 'gray')
```

```python
plt.figure(figsize =(16,9))
plt.subplot(1,3,1)
plt.title('Orignal Image')
plt.imshow(img, cmap = 'gray')

plt.subplot(1,3,2)
plt.title('Sobel')
plt.imshow(binary_sobel, cmap = 'gray')

plt.subplot(1,3,3)
plt.title('Canny')
plt.imshow(edges, cmap = 'gray')
```

```python
img = cv2.imread('circle.png', 0)
```

```python
edges = cv2.Canny(img, 60, 255)

sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = 3)
final_sobel = cv2.add(sobelx, sobely)
thresh, binary_sobel = cv2.threshold(final_sobel, 60, 255, cv2.THRESH_BINARY)


plt.figure(figsize =(16,9))
plt.subplot(1,3,1)
plt.title('Orignal Image')
plt.imshow(img, cmap = 'gray')

plt.subplot(1,3,2)
plt.title('Sobel')
plt.imshow(binary_sobel, cmap = 'gray')

plt.subplot(1,3,3)
plt.title('Edge Detected')
plt.imshow(edges, cmap = 'gray')
```
