```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("circle.png", 0)
```

```python
plt.imshow(img, cmap = "grey")
```

```python
help(cv2.cornerHarris)
```

```python
corners = cv2.cornerHarris(img, blockSize = 3, ksize = 3, k = 0.04)
```

```python
corners
```

```python
corners.shape
```

```python
plt.imshow(corners, cmap = "grey")
```

```python
new = corners > corners.max() / 2
plt.imshow(new, cmap = "grey")
```

```python
new2 = corners  < corners.max() / 2
plt.imshow(new2, cmap = "grey")
```

```python
lenna = cv2.imread("lenna.png", 0)
```

```python
lenna_corners = cv2.cornerHarris(lenna, blockSize = 3, ksize = 3, k = 0.04)
new_corns = lenna_corners > lenna_corners.max() / 2
plt.imshow(new_corns, cmap = "grey")
```
