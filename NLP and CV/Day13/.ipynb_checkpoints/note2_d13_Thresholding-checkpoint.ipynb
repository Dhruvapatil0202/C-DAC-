```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sachin.jpg", 0)
```

##### Binary Thresholding

```python
binary_image = img > 127
binary_image
```

```python
binary_image.apply(lambda x : 255 if x else 0)
```

```python
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, cmap = "gray")
plt.subplot(1, 2, 2)
plt.title("Binary")
plt.imshow(binary_image, cmap = 'gray')
```

```python
lenna = cv2.imread("lenna.png", 0)
binary_lenna = lenna > 127

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(lenna, cmap = "gray")
plt.subplot(1, 2, 2)
plt.title("Binary")
plt.imshow(binary_lenna, cmap = 'gray')
```

```python
lenna = cv2.imread("lenna.png", 0)
binary_lenna = lenna > 127

new_lenna = binary_lenna.astype(np.uint8) * 255

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(lenna, cmap = "gray")
plt.subplot(1, 2, 2)
plt.title("Binary")
plt.imshow(new_lenna, cmap = 'gray')
```

```python
lenna = cv2.imread("lenna.png", 0)

thresh, new_lenna = cv2.threshold(lenna, 127, 255, cv2.THRESH_BINARY)
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(lenna, cmap = "gray")
plt.subplot(1, 2, 2)
plt.title("Binary")
plt.imshow(new_lenna, cmap = 'gray')
```

```python
lenna = cv2.imread("lenna.png", 0)
binary_lenna = lenna < 127

new_lenna = binary_lenna.astype(np.uint8) * 255

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(lenna, cmap = "gray")
plt.subplot(1, 2, 2)
plt.title("Inverse Binary")
plt.imshow(new_lenna, cmap = 'gray')
```

##### Otsu Thresholding

```python
lenna = cv2.imread("lenna.png", 0)

thresh, new_lenna_otsu = cv2.threshold(lenna, 100, 255, cv2.THRESH_OTSU)

thresh, new_lenna_binary = cv2.threshold(lenna, 100, 255, cv2.THRESH_BINARY)

plt.figure(figsize = (16, 9))

plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(lenna, cmap = "gray")

plt.subplot(1, 3, 2)
plt.title("Otsu")
plt.imshow(new_lenna_otsu, cmap = 'gray')

plt.subplot(1, 3, 3)
plt.title("Binary")
plt.imshow(new_lenna_binary, cmap = 'gray')
```
