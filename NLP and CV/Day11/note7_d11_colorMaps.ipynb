```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sachin.jpg', 0)
```

```python
img.shape
```

```python
plt.imshow(img, cmap="Blues")
```

```python
orig_map = plt.colormaps['Blues']
reversed_map = orig_map.reversed()
plt.imshow(img, cmap = reversed_map)
```

```python
plt.figure(figsize = (16,8))
plt.subplot(4,4,1);  plt.title('Blues')
plt.imshow(img, cmap = plt.colormaps['Blues'].reversed())
plt.subplot(4,4,2);  plt.title('Greens')
plt.imshow(img, cmap = plt.colormaps['Greens'].reversed())
plt.subplot(4,4,3);  plt.title('Reds')
plt.imshow(img, cmap = plt.colormaps['Reds'].reversed())
plt.subplot(4,4,4);  plt.title('Oranges')
plt.imshow(img, cmap = plt.colormaps['Oranges'].reversed())
plt.subplot(4,4,5);  plt.title('Purples')
plt.imshow(img, cmap = plt.colormaps['Purples'].reversed())
plt.subplot(4,4,6);  plt.title('Greys')
plt.imshow(img, cmap = plt.colormaps['Greys'].reversed())
plt.subplot(4,4,7);  plt.title('YlOrBr')
plt.imshow(img, cmap = plt.colormaps['YlOrBr'].reversed())
plt.subplot(4,4,8);  plt.title('OrRd')
plt.imshow(img, cmap = plt.colormaps['OrRd'].reversed())

plt.subplot(4,4,9);  plt.title('YlOrRd')
plt.imshow(img, cmap = plt.colormaps['YlOrRd'].reversed())
plt.subplot(4,4,10);  plt.title('OrRd')
plt.imshow(img, cmap = plt.colormaps['OrRd'].reversed())
plt.subplot(4,4,11);  plt.title('PuRd')
plt.imshow(img, cmap = plt.colormaps['PuRd'].reversed())
plt.subplot(4,4,12);  plt.title('RdPu')
plt.imshow(img, cmap = plt.colormaps['RdPu'].reversed())
plt.subplot(4,4,13);  plt.title('BuPu')
plt.imshow(img, cmap = plt.colormaps['BuPu'].reversed())
plt.subplot(4,4,14);  plt.title('GnBu')
plt.imshow(img, cmap = plt.colormaps['GnBu'].reversed())
plt.subplot(4,4,15);  plt.title('BuPu')
plt.imshow(img, cmap = plt.colormaps['BuPu'].reversed())
plt.subplot(4,4,16);  plt.title('YlGn')
plt.imshow(img, cmap = plt.colormaps['YlGn'].reversed())

plt.tight_layout()
```
