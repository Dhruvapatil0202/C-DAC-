```python
from urllib.request import urlopen
```

```python
url = urlopen('https://en.wikipedia.org/wiki/Rajgad_Fort')
```

```python
html = url.read()
```

```python
html
```

```python
# beautify by removing all the tags in the above data
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
soup.find('title')
```

```python
soup.find('h1')
```

```python
data = soup.get_text()
```

```python
print(data)
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(data)
```

```python
tokens
```

```python
from nltk import FreqDist
```

```python
freq = FreqDist(tokens)
```

```python
freq.plot(20)
```

```python
# pip install matplotlib
```

```python
from nltk.corpus import stopwords
```

```python
tokens1 = [token.lower() for token in tokens if token.isalpha() or token.isdigit()]
```

```python
tokens1
```

```python
freq = FreqDist(tokens1)
freq.plot(25)
```

```python
s_words = stopwords.words("english")
tokens2 = [word.strip() for word in tokens1 if word not in s_words]
freq = FreqDist(tokens2)
freq.plot(25)
```

```python
freq.most_common(5)
```

```python
from nltk.stem import PorterStemmer
ls = PorterStemmer()
```

```python
url = urlopen("https://en.wikipedia.org/wiki/Tiger")
data = url.read()
soup = BeautifulSoup(data, "html.parser")
text = soup.get_text()
tokens = word_tokenize(text)
swords = stopwords.words("english")
tokens1 = [token.lower() for token in tokens if token.isalpha() or token.isnumeric()]
tokens2 = [ls.stem(token) for token in tokens1 if token not in swords and len(token) > 1]
freq  = FreqDist(tokens2)
freq.plot(20)
freq.most_common(5)
```
