##### cleaning using regular expression

```python
import re
import string

def cleanResume(resumeText):
    resumeText = re.sub("http\S+\s*", " ", resumeText) # remove URLs
    resumeText = re.sub("RT|cc", " ", resumeText) # remove RT and cc
    resumeText = re.sub("#\S+", " ", resumeText) # remove hashtags
    resumeText = re.sub("@\S+", " ", resumeText) # remove mentions
    resumeText = re.sub("[%s]"% re.escape(string.punctuation), " ", resumeText) # Remove punctuations
    resumeText = re.sub(r"[^\x00-\x7f]", r" ", resumeText)
    resumeText = re.sub("\s+", " ", resumeText) # remove extra whitespace
    return resumeText
```

```python
cleanedResume = cleanResume(df.iloc[0]["Resume"])
cleanedResume
```

```python
cleanData = ""
for index, row in df.iterrows():
    cleanData += cleanResume(row["Resume"])
cleanData
```

```python
cleaned = x.apply(cleanResume)
cleaned
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words = "english")
```

```python
x_new = tfidf.fit_transform(cleaned)
x_new.shape
```

```python
tfidf.get_feature_names_out()
```

##### cross validation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_new, y, random_state = 0)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.neighbors import NearestCentroid
```

```python
nc = NearestCentroid()
nc.fit(x_train, y_train)
```

##### Performance Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay, confusion_matrix
```

```python
y_pred = nc.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize = (25, 20))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```
