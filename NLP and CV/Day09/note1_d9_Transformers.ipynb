# Question - Answering

```python
qa = pipeline('question-answering')
```

```python
text = '''Attack on Titan (AOT) Season 1 follows Eren Yeager, his adoptive sister Mikasa Ackerman, and their friend Armin Arlert, as they fight for survival in a world where humanity is on the brink of extinction. The story takes place within enormous walled cities that protect the last remnants of humanity from giant humanoid creatures known as Titans, which devour humans.

The season begins with a devastating attack by the Titans, led by the appearance of the Colossal Titan, which breaches Wall Maria, allowing smaller Titans to invade. Eren's mother is killed during the attack, sparking his desire for revenge. Eren, Mikasa, and Armin enlist in the military to fight against the Titans, joining the elite Survey Corps, which ventures outside the walls to battle these terrifying creatures.

As the series progresses, the trio uncovers shocking truths about the Titans, including Eren's own mysterious ability to transform into a Titan. They encounter powerful enemies such as the Armored Titan and the Female Titan, who seem to have a hidden agenda. The season explores themes of survival, sacrifice, and the blurred lines between good and evil. The shocking revelations about the Titans set the stage for the larger mysteries to unfold in future seasons.'''
```

```python
question = 'Who are Titans?'
```

```python
result = qa(question = question, context = text)
df = pd.DataFrame([result])
df
```

```python
question2 = 'What is Eren Yeagers mysterous ability?'
result = qa(question = question2, context = text)
df = pd.DataFrame([result])
df
```

```python
question3 = 'Who attacked wall Maria?'
result = qa(question = question3, context = text)
df = pd.DataFrame([result])
df
```

# Visual Question Answering

```python
from PIL import Image
```

```python
vqa = pipeline('visual-question-answering')
```

```python
image = Image.open('earth.jpg')
```

```python
question = 'Who is in the image?'
vqa(image, question, top_k=31)[30]['answer']
```

```python
question = 'What star is doing?'
vqa(image, question, top_k=31)[30]['answer']
```

```python
question = 'Where star is sitting?'
vqa(image, question, top_k=31)[30]['answer']
```

```python
image2 = Image.open('tiger.jpg')
```

```python
question = 'Who is in the image?'
vqa(image2, question, top_k=5)
```

```python
question = 'Who is in the image?'
vqa(image2, question, top_k=5)[4]['answer']
```

```python
question = 'What tiger is doing?'
vqa(image2, question, top_k=3)[0]['answer']
```

```python
question = 'Where tiger is sitting?'
vqa(image2, question, top_k=3)[1]['answer']
```

```python
question = 'What is the color of the tiger?'
vqa(image2, question, top_k=3)[0]['answer']
```
