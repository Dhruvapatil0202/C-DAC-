```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sachin.jpg")
```

```python
dim = 11
kernal = np.array([[1 / (dim * dim)] * dim] * dim)

# kernal = np.ones([5, 5]) * 1 / 25
```

```python
kernal
```

help(cv2.filter2D)

```python
smoothened_image = cv2.filter2D(img, -1, kernal)
```

```python
cv2.imshow("Blurred Image", smoothened_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Gaussian Kernal

```python
kernal_size = 5
sigma = 0.35 *  ((kernal_size-1)*0.5 -1) + 0.9 
kernal = cv2.getGaussianKernel(kernal_size, sigma)
```

```python
kernal
```

```python
sigma
```

```python
gauss_image = cv2.filter2D(img, -1, kernal)
```

```python
cv2.imshow('Blurred Image', gauss_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_img = np.hstack([img, gauss_image])

cv2.imshow('Image', new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Median Blur

```python
gray = cv2.imread('sachin.jpg', 0)
```

```python
gray.shape
```

```python
img_med = cv2.medianBlur(gray, 5) # 5 ---> kernel size
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Orignal image')
plt.imshow(gray, cmap='gray')
plt.subplot(1,2,2)
plt.title('Blurred image')
plt.imshow(img_med, cmap='gray');
```

```python
sub_img = cv2.imread('sachin.jpg', 0)
```

```python
sub_img = sub_img[200:400,400:700]
```

```python
sub_img.shape
```

```python
img_med = cv2.medianBlur(sub_img, 11) # 5 ---> kernel size
```

```python
new_img = cv2.imread('sachin.jpg', 0)
```
